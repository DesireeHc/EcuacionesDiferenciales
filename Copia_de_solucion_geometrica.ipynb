## Problema

Una fábrica produce dos tipos de recipientes, K y L, para lo cual utiliza dos máquinas, $M_1$ y $M_2$. Para producir un recipiente K, $M_1$ necesita 2 minutos y $M_2$, 4 minutos. De manera semejante, L ocupa 8 minutos a $M_1$ y 4 minutos a $M_2$. La ganancia neta para un recipiente K es de 29 pesos y para uno L es de 45 pesos. Determinar el plan de producción que maximice la ganancia neta.

De manera formal:

#### Variables de decisión:

Número de recipientes K, $x_1$ y L, $x_2$.

#### Función Objetivo:

Maximizar $29x_1 + 45x_2$

#### Restricciones:

Consideramos la producción por hora:

Resultante de la máquina $M_1$: $2x_1 + 8x_2 \leq 60$

Resultante de la máquina $M_2$: $4x_1 + 4x_2 \leq 60$

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

Importamos el módulo pyplot de la biblioteca matplotlib, asignandole el nombre plt


```python
import matplotlib.pyplot as plt #Permite crear gráficos, figuras, ejes, puntos, etc
%matplotlib inline
#Se usa para que los graficos generados anteriormente se muestren directamente en el notebook, facilita análisis de datos
```

```python
# Crea una nueva figura para dibujar el gráfico
plt.figure()
# Al crear una nueva figura con plt.figure(), los gráficos se dibujan en ventanas separadas, (no sobreescribir gráficos anteriores o al trabajar con varios gráficos de forma independiente)


# --- Restricción para M1 ---
# Dibuja una línea que representa la restricción M1
# Los puntos (0, 7.5) y (30, 0) se conectan con una línea
plt.plot([0, 60.0/2.0],                # Valores del eje X (de 0 a 30) se obtienen de la restricción M1
         [60.0/8.0, 0],                # Valores del eje Y (de 7.5 a 0) se obtienen de la restricción M1
         color='orange',               # Establece el color de la línea en naranja
         alpha=0.25)                   # Establece la transparencia de la línea (25%)

# Rellena el área bajo la línea que se ha trazado para M1
plt.fill_between([0, 60.0/2.0],        # Valores del eje X
                 [60.0/8.0, 0],        # Valores del eje Y
                 color='orange',        # Color del área rellena
                 alpha=0.25,            # Transparencia del área rellena
                 label='resultante de M1')  # Etiqueta para la leyenda
# plt.fill_between() rellena el área entre la línea que hemos trazado y el eje X.


# --- Restricción para M2 ---
# Dibuja una línea que representa la restricción M2
# Los puntos (0, 15) y (15, 0) se conectan con una línea
plt.plot([0, 60.0/4.0],                # Valores del eje X (de 0 a 15) se obtienen de la restricción M2
         [60.0/4.0, 0],                # Valores del eje Y (de 15 a 0) se obtienen de la restricción M2
         color='blue',                 # Establece el color de la línea en azul
         alpha=0.125)                  # Establece la transparencia de la línea (12.5%)

# Rellena el área bajo la línea que se ha trazado para M2
plt.fill_between([0, 60.0/4.0],        # Valores del eje X (igual que antes)
                 [60.0/4.0, 0],        # Valores del eje Y (igual que antes)
                 color='blue',          # Color del área rellena
                 alpha=0.125,           # Transparencia del área rellena
                 label='resultante de M2')  # Etiqueta para la leyenda

# Etiqueta el eje X
plt.xlabel('recipientes K')             # Describe qué representa el eje X

# Etiqueta el eje Y
plt.ylabel('recipientes L')             # Añade una etiqueta al eje Y

# Muestra la leyenda del gráfico
plt.legend()                             # Incluye la leyenda que explica el gráfico,  permite visualizar el significado de cada elemento

# Muestra el gráfico en pantalla, ultima acción para visualizar el gráfico completo con todos los elementos añadidos.
plt.show()                               # Genera y muestra el gráfico
```

### Ejercicio 3:
Juan debe trabajar cuando menos 20 horas a la semana para complementar sus
ingresos, y al mismo tiempo asistir a la escuela. Tiene la oportunidad de trabajar
en dos tiendas al menudeo: en la tienda 1 puede trabajar entre 5 y 12 horas por
semana, y en la tienda 2 le permiten trabajar entre 6 y 10 horas. Ambas tiendas le
pagan el mismo sueldo por hora. En consecuencia, Juan quiere basar su decisión
acerca de cuántas horas trabajar en cada tienda en un criterio distinto: el factor
de tensión en el trabajo. Con base en las entrevistas con otros empleados, Juan
estima que en una escala de 1 a 10, los factores de tensión son 8 y 6 en las tiendas
1 y 2, respectivamente. Como la tensión aumenta cada hora, supone que la tensión
total al final de la semana es proporcional a la cantidad de horas que trabaja en
las tiendas. ¿Cuántas horas debería trabajar Juan en cada tienda?


De manera formal:

#### Variables de decisión:

$x_1$ y $x_2$ son las horas que debe decidir trabajar en cada tienda (tienda 1 y 2 respectivamente).

#### Función Objetivo (minimizar la tensión total):


Minimizar  $Z=8x_1+6x_2$


#### Restricciones:

* $x_1+x_2 ≥20$ (debe trabajar al menos 20 horas).
* $5≤x_1 ≤12$ (horas en la tienda 1).
* $6≤x_2≤10$ (horas en la tienda 2).

No negatividad: $x_1 \geq 0$, $x_2 \geq 0$

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure()

# Restricción 1
plt.plot([0, 20], [20, 0], color='orange', alpha=0.25)
plt.fill_between([0, 20], [20, 0], color='orange', alpha=0.25, label='resultante de 1')

x = (0, 30)
y = (0, 30)

# Restricción 2
x1 = 5
x2 = 12

plt.axvline(x=x1, color='blue', alpha=0.125, label='Restricción 2 (x1)')
plt.axvline(x=x2, color='blue', alpha=0.125, label='Restricción 2 (x2)')
plt.fill_between([x1, x2], 0, 20, color='lightblue', alpha=0.7)

# Restricción 3

y1 = 6
y2= 10
plt.axhline(y=y1, color='blue', alpha=0.125, label='Restricción 2 (y1)')
plt.axhline(y=y2, color='blue', alpha=0.125, label='Restricción 2 (y2)')
plt.fill_between(x, y1, y2, color='blue', alpha=0.125)

plt.xlabel('horas trabajadas en tienda 1')
plt.ylabel('horas trabajadas en tienda 2')

plt.scatter([12], [8])  # punto factible
plt.plot([0, 144.0/8], [144.0/6, 0], color='black', linestyle='--')

plt.scatter([10], [10], color='red')  # punto factible
plt.plot([0, 140/8], [140/6, 0], color='red', linestyle='-.')


plt.xlim(0)
plt.ylim(0)

plt.legend()
plt.show()
```
