#Modelo de Lucha entre dos Grupos (Hirschman)


El Modelo de Lucha entre dos Grupos, también conocido como __el Modelo de Hirschman__, fue desarrollado por el economista Albert O. Hirschman en 1971. Este modelo matemático describe la dinámica del conflicto entre dos grupos sociales, considerando su __crecimiento y la intensidad de la lucha entre ellos__. A través de un sistema de ecuaciones diferenciales, el modelo analiza la _evolución de las poblaciones de ambos grupos a lo largo del tiempo_.

 El modelo se enmarca en un contexto histórico marcado por la __creciente atención a los conflictos sociales y la búsqueda de herramientas para comprenderlos y abordarlos__. La década de 1960 y principios de la de 1970 estuvieron marcadas por un período de gran agitación social y política en todo el mundo. Movimientos como la lucha por los derechos civiles en los Estados Unidos, la oposición a la Guerra de Vietnam y el surgimiento de movimientos estudiantiles radicales pusieron de relieve las tensiones y conflictos existentes en las sociedades.  Gracias a esto, surgió una necesidad creciente de comprender sus causas, dinámicas y posibles soluciones. Los científicos sociales comenzaron a __explorar diversos enfoques teóricos para explicar el comportamiento de los grupos en conflicto__ y las condiciones que favorecían la violencia o la cooperación.

 La Teoría de Juegos, desarrollada a mediados del siglo XX, proporcionó un marco formal para analizar las interacciones estratégicas entre actores. Hirschman se inspiró en este enfoque para modelar las decisiones de los grupos en conflicto, considerando los beneficios y costos de diferentes estrategias.

##¿Qué es?

El modelo de Hirschman se centra en el __estudio de la dinámica del conflicto entre dos grupos__, examinando cómo el crecimiento y la lucha entre ellos afectan las poblaciones de cada grupo a lo largo del tiempo.

##¿Para qué?

El modelo tiene como objetivo __comprender las causas y consecuencias del conflicto entre grupos__, identificar __patrones__ en la dinámica de estos conflictos y explorar __estrategias para manejarlos o prevenirlos__.

El modelo es aplicable a una amplia gama de situaciones que involucran el conflicto entre dos grupos, como _conflictos étnicos, políticos, económicos, sociales, entre otros_.



## Suposiciones:

###Crecimiento Independiente:

 Cada grupo crece a una tasa constante en ausencia de conflicto. Esta tasa de crecimiento está representada por los parámetros $r$ para el Grupo 1 y $s$ para el Grupo 2.

###Reducción del Crecimiento Inducido por el Conflicto:
 El conflicto entre los dos grupos reduce la tasa de crecimiento de cada grupo. Esta reducción es proporcional al tamaño del otro grupo y está representada por los coeficientes $a$ para el Grupo 1 y $b$ para el Grupo 2.

###Dinámica de Tiempo Continuo:

El modelo asume que los procesos de crecimiento y conflicto ocurren continuamente en el tiempo. Esto permite una representación dinámica de los cambios en el tamaño de los dos grupos.

### Derivación de Ecuaciones:

Basándonos en estas suposiciones, podemos expresar la tasa de cambio de la población de cada grupo a lo largo del tiempo utilizando ecuaciones diferenciales. La __tasa de cambio de una población es la derivada de su tamaño con respecto al tiempo__.

*Grupo 1:*

La tasa de cambio de la población del Grupo 1 (X(t)) se puede expresar como:

$$\frac{dX}{dt} = \frac{d(X(t))}{dt}$$

Esta tasa de cambio está influenciada por dos factores:

1. Crecimiento Independiente: En ausencia de conflicto, el Grupo 1 crecería a una tasa de r.

2. Reducción del Crecimiento Inducido por el Conflicto: El conflicto con el Grupo 2 reduce la tasa de crecimiento del Grupo 1 en una cantidad proporcional al tamaño del Grupo 2 (Y(t)) y al coeficiente de conflicto a.

Combinando estos factores, obtenemos la siguiente ecuación para la tasa de cambio de la población del Grupo 1:

$$\frac{dX}{dt} = rX - aXY$$

_Grupo 2:_

De manera similar, la tasa de cambio de la población del Grupo 2 (Y(t)) se puede expresar como:

$$\frac{dY}{dt} = \frac{d(Y(t))}{dt}$$

Esta tasa de cambio también está influenciada por dos factores:

1. Crecimiento Independiente: En ausencia de conflicto, el Grupo 2 crecería a una tasa de s.

2. Reducción del Crecimiento Inducido por el Conflicto: El conflicto con el Grupo 1 reduce la tasa de crecimiento del Grupo 2 en una cantidad proporcional al tamaño del Grupo 1 (X(t)) y al coeficiente de conflicto b.

Combinando estos factores, obtenemos la siguiente ecuación para la tasa de cambio de la población del Grupo 2:

$$\frac{dY}{dt} = sY - bXY$$

## Ecuaciones del modelo:

El modelo de Hirschman se representa mediante el sistema de ecuaciones diferenciales de primer orden:

$$
\left\{
\begin{array}{ll}
\frac{dX}{dt} = rX - aXY \\
\frac{dY}{dt} = sY - bXY
\end{array}
\right.
$$

Donde:

* $X(t)$ representa el tamaño del grupo 1 en el momento t.
* $Y(t)$ representa el tamaño del grupo 2 en el momento t.
* $r$ es la tasa de crecimiento natural del grupo 1.
* $s$ es la tasa de crecimiento natural del grupo 2.
* $a$ es el coeficiente de conflicto para el grupo 1.
* $b$ es el coeficiente de conflicto para el grupo 2.



##Puntos críticos:


Los puntos críticos del modelo son los puntos en el plano (X, Y) donde las tasas de cambio de ambas poblaciones son iguales a cero:

$\frac{dX}{dt} = 0$

$\frac{dY}{dt} = 0$

Resolviendo este sistema de ecuaciones, se obtienen tres puntos críticos:

* $(0, 0)$: Equilibrio de extinción, donde ambos grupos se extinguen.

* $(K1, 0)$: Equilibrio de coexistencia, donde el grupo 1 persiste y el grupo 2 se extingue.

* $(0, K2)$: Equilibrio de coexistencia, donde el grupo 2 persiste y el grupo 1 se extingue.

##Clasificación de los puntos críticos:

La estabilidad de los puntos críticos se determina analizando los eigenvalores de la matriz jacobiana del sistema de ecuaciones en cada punto crítico.

* Equilibrio de extinción (0, 0): Este equilibrio es siempre inestable.
* Equilibrio de coexistencia (K1, 0): Este equilibrio es estable si $rs - abK1K2 > 0$ e inestable si $rs - abK1K2 < 0$.
* Equilibrio de coexistencia (0, K2): Este equilibrio es estable si $rs - abK1K2 > 0$ e inestable si $rs - abK1K2 < 0$.


##Retrato fase:

El retrato fase del modelo de Hirschman __representa las diferentes regiones del plano (X, Y) donde el sistema se comporta de manera similar__. Las trayectorias del sistema muestran la evolución de las poblaciones de los grupos a lo largo del tiempo. El análisis del retrato fase permite __identificar las regiones de atracción hacia los equilibrios estables y las regiones de inestabilidad__.


##Teoremas accesorios:

###Condiciones de existencia y unicidad:

Las condiciones para la existencia y unicidad de soluciones en el modelo de Hirschman no son triviales y dependen de los valores de los parámetros (r, s, a, b). En general, se requiere que los parámetros sean positivos, esta condición asegura que las __tasas de crecimiento y los coeficientes de conflicto sean valores significativos__. y que las tasas de crecimiento naturales (r, s) sean mayores que los coeficientes de conflicto (a, b), permitiendo que las poblaciones puedan crecer en ausencia de conflicto.

##Cantidades conservadas o acotadas:

En el modelo de Hirschman, __no hay variables que mantengan un valor constante a lo largo del tiempo__. Sin embargo, existen ciertas relaciones entre las variables que se mantienen constantes o acotadas:

* Suma total de poblaciones:

La suma total de las poblaciones de los dos grupos siempre será mayor o igual que cero:
$X(t) + Y(t) >= 0$
Esta relación implica que la cantidad total de individuos en el sistema siempre será no negativa.

* Proporción relativa de poblaciones:

La razón entre el tamaño del grupo 1 y el tamaño del grupo 2, no está acotada. Esta proporción puede variar sin límites a lo largo del tiempo, dependiendo de la dinámica del conflicto entre los grupos.


## Importancia de los Teoremas Accesorios:

Estas condiciones permiten establecer límites a las posibles soluciones del modelo y restringen el rango de valores que pueden tomar las variables (X, Y) a lo largo del tiempo.

La ausencia de cantidades conservadas en el modelo de Hirschman indica que la dinámica del conflicto entre los grupos puede ser _compleja e impredecible_. La proporción relativa de las poblaciones puede variar sin límites, lo que significa que un grupo puede dominar al otro o ambos grupos pueden coexistir en diferentes proporciones.


```python
# Importamos las librerias necesarias
import sympy as sym  # Para cálculos simbólicos
import matplotlib.pyplot as plt  # Para gráficos
import numpy as np  # Para cálculos numéricos
from scipy.integrate import odeint  # Para integrar ecuaciones diferenciales ordinarias

# Definir símbolos y parámetros
x, y = sym.symbols('x y')
a, b, c, d = 1.2, 0.5, 1.1, 0.6 # Valores de ejemplo para los parámetros

# Definir las ecuaciones del modelo de Lucha entre dos Grupos (Hirschman)
dx_dt = a * x - b * x * y
dy_dt = -c * y + d * x * y

# Encontrar los puntos críticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")

# Retrato fase del modelo
def hirschman(X, t, a, b, c, d):
    x, y = X
    dx_dt = a * x - b * x * y
    dy_dt = -c * y + d * x * y
    return [dx_dt, dy_dt]

x_range = np.linspace(0, 10, 20)
y_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Grupo 1 (x)')
plt.ylabel('Grupo 2 (y)')
plt.title('Retrato Fase del Modelo de Hirschman')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Hirschman
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 4)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(func, x0, ts, args=(a, b, c, d))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Grupo 1 (x)', fontsize=10)
    plt.ylabel('Grupo 2 (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 10)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:10:10j, 0:10:10j]
    u = a * X - b * X * Y
    v = -c * Y + d * X * Y
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(hirschman)
```

## Interpretación

Las trayectorias resultantes representan los caminos que toma el sistema a medida que __cambian las poblaciones de los dos grupos__.

El retrato fase para este escenario muestra dos equilibrios estables, el de extinción y el de coexistencia.

Esto indica que si las poblaciones iniciales de los dos grupos están cerca del equilibrio de extinción, _ambos se extinguirán_. Sin embargo, si las poblaciones iniciales están cerca del equilibrio de coexistencia, los dos grupos _coexistirán y mantendrán poblaciones relativamente estables_.
