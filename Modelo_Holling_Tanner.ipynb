#Modelo de Holling Tanner



La relaci√≥n entre depredadores y presas es un _fen√≥meno natural de la din√°mica poblacional_, rama de la ecolog√≠a, llamado _depredaci√≥n_. En este tipo de fen√≥menos existen escenarios  tales como el __efecto indirecto de un depredador sobre la presa__, en este, basta la presencia del depredador, a√∫n sin haber depredaci√≥n, para que el __comportamiento y densidad de poblaci√≥n  de la presa se vean afectados__. Este tipo de escenarios se pueden modelar con un sistema depredador-presa donde se busca representar dichos efectos indirectos.



El modelo fue desarrollado por __Crawford Stanley Holling__ y __John T. Tanner__. Holling, un ec√≥logo canadiense, es conocido por su trabajo en la respuesta funcional de los depredadores. _Introdujo la idea de diferentes tipos de respuestas funcionales_, siendo la Tipo II una de las m√°s estudiadas. John T. Tanner, un matem√°tico y ec√≥logo, contribuy√≥ al desarrollo del modelo matem√°tico en 1970. Su trabajo ayud√≥ a _formalizar la din√°mica poblacional de depredadores y presas usando ecuaciones diferenciales_. Tanner se centr√≥ en la aplicaci√≥n de estos modelos para _comprender la estabilidad y las fluctuaciones c√≠clicas en las poblaciones naturales_.

Este modelo ha sido utilizado para estudiar sistemas _depredador-presa_, reproduciendo patrones observados en sistemas poblacionales, tales como _ciclos poblacionales_, _oscilaciones ca√≥ticas_ y _extinciones locales_.

Los modelos depredador-presa se clasifican en tres tipos, __los que dependen de la presa, los dependientes del depredador, o los mixtos__. Estas clasificaciones dependen de la rapidez con la que los depredadores eliminan a sus presas. Hollig propone tres tipos de respuestas funcionales, de los cuales, en su mayoria son aplicados a modelos que dependen de la presa.

_aumento lineal de la tasa de ataque de la poblacion de depredadores respecto a la densidad de poblacion de presas_, hasta llegar a un punto a

##¬øQu√©?

El modelo de Holling-Tanner es un __sistema de ecuaciones diferenciales__ que describe la __din√°mica de una poblaci√≥n de presas y una poblaci√≥n de depredadores__. Este modelo tiene en cuenta una respuesta funcional tipo II, que es una representaci√≥n m√°s realista de la tasa de depredaci√≥n en comparaci√≥n con otros modelos, como el modelo Lotka-Volterra.




##¬øPara qu√©?

El modelo se utiliza para entender __c√≥mo interact√∫an las poblaciones de presas y depredadores en un ecosistema__. Ayuda a prever la _estabilidad de las poblaciones_ y las condiciones bajo las cuales las poblaciones pueden coexistir de manera sostenible. Este conocimiento es crucial en la gesti√≥n de la vida silvestre, la conservaci√≥n de especies y el manejo de recursos naturales.





##¬øQui√©n?
El modelo es de inter√©s para ec√≥logos, bi√≥logos, conservacionistas y gestores de recursos naturales. Investigadores acad√©micos y estudiantes de biolog√≠a matem√°tica tambi√©n utilizan este modelo para estudiar la din√°mica de poblaciones.



##Ecuaciones

Estudiaremos la respuesta funcional Holling tipo II, el cual
tiene la siguiente forma:

$$h(x)=\frac{qx}{x+a}$$

donde $q>0$ y $a>0$
y $q$ es la tasa m√°xima de crecimiento del depredador y $a$ es la medida de protecci√≥n a la presa brindada por el medio ambiente.

Sin embargo, muchos bi√≥logos han cuestionado que este modelo solo dependa de la presa, por lo que se propone una respuesta funcional mixta, donde se dependa tanto de la presa como del depredador, a esto se le llama _raz√≥n- dependiente_, as√≠ se utilizar√°:

$$h(x)=\frac{qx}{x+ay}$$

En cualquier tiempo t, la __poblaci√≥n de las presas__ se denota como $x$, mientras que la __poblaci√≥n de los depredadores__ se denotar√°n como $y$. En este caso en particular, la poblaci√≥n de la presa es mucho mayor que la del depredador, este modelo est√° dado por el siguiente sistema de ecuaciones:

$$\begin{cases}
\frac{dx}{dt}=r(1-\frac{x}{K}-\frac{q}{x+ay}y)x \\
\frac{dy}{dt}=s(1-\frac{y}{nx})y
\end{cases}
$$

##Par√°metros


* $x$ : Poblaci√≥n de presas.

* $y$ : Poblaci√≥n de depredadores.


* $r$ : Tasa de crecimiento intr√≠nseco para la presa (sin restricciones).

 * $0<r‚â§1$.
 Este par√°metro describe la _velocidad a la cual la poblaci√≥n de presas crece_ en ausencia de depredadores.


* $s$ : Tasa de crecimiento intr√≠nseco para el depredador


* $K$ : Capacidad de carga del entorno para la poblaci√≥n de presas.

 * $K>0$.
Este par√°metro representa el _tama√±o m√°ximo de la poblaci√≥n de presas que el ambiente puede soportar_.

* $a$ : Tasa m√°xima de depredaci√≥n.

 * $a>0$.
Este par√°metro indica la tasa a la cual los _depredadores consumen presas cuando las presas son abundantes_.

* $n$ : Medida de calidad de la presa como alimento para el depredador.

 * $q$ y $n$ deben ser __menores que uno__ para reflejar que __no todas las presas consumidas se convierten directamente en nuevos depredadores__; parte de la energ√≠a se pierde en el proceso.


##Interpretaci√≥n de los T√©rminos en las Ecuaciones



###Ecuaci√≥n de la Poblaci√≥n de Presas

$$\frac{dx}{dt}=r(1-\frac{x}{K}-\frac{q}{x+ay}y)x$$‚Äã



El t√©rmino $1-\frac{x}{K}$ representa el __crecimiento log√≠stico de las presas__, limitado por la __capacidad de carga__ $k$. El t√©rmino $\frac{q}{x+ay}y$ representa la __tasa de depredaci√≥n __ajustada por la __disponibilidad de presas__ y la __competencia entre depredadores__, el denominador, $x+ay$   representa la __dificultad de capturar presas__ a medida que aumentan las interacciones presa-depredador.

###Ecuaci√≥n de la Poblaci√≥n de Depredadores

$$\frac{dy}{dt}=s(1-\frac{y}{nx})$$


El t√©rmino $1-\frac{y}{nx}$
  representa el __crecimiento de la poblaci√≥n de depredadores__, limitado por la disponibilidad de presas. Si la cantidad de presas es insuficiente (
$y‚â´nx$), la poblaci√≥n de depredadores disminuye. Si hay suficientes presas (
$$y‚â™nx$$), la poblaci√≥n de depredadores puede crecer.

###Din√°mica del Sistema


__Presas ($x$):__

* Crecen log√≠sticamente con una capacidad de carga
$K$.

* Son depredadas por los depredadores $y$, con una _tasa de depredaci√≥n que sigue una respuesta funcional tipo II de Holling_, donde la tasa de depredaci√≥n se estabiliza a altos niveles de presas debido a la saturaci√≥n del depredador.


__Depredadores ($y$):__

* Crecen en funci√≥n de la disponibilidad de presas $x$.
* La tasa de crecimiento de los depredadores disminuye si la poblaci√≥n de presas es insuficiente para sostener su crecimiento.

### Puntos criticos

Los puntos cr√≠ticos representan los estados estacionarios del sistema, donde las __tasas de cambio de las poblaciones de presas y depredadores son cero__. Es decir, son las __soluciones donde las poblaciones de presas y depredadores no cambian con el tiempo__.

* __Punto cr√≠tico (0, 0)__: Ambas poblaciones est√°n extintas. No hay presas ni depredadores presentes en el sistema.

* __Punto cr√≠tico donde
$ùë•‚â†0$ y $y‚â†0$__: Este punto cr√≠tico se determina resolviendo el sistema de ecuaciones igualado a cero.


__Por qu√© son importantes los puntos cr√≠ticos:__

Los puntos cr√≠ticos son fundamentales para entender las din√°micas del sistema predador-presa, la estabilidad de estos puntos cr√≠ticos nos ayuda a prever __c√≥mo responder√° el sistema ante cambios en los par√°metros del modelo o en las condiciones ambientales__.



##Teoremas Accesorios

###Existencia y Unicidad
La existencia y unicidad de soluciones para el sistema de ecuaciones diferenciales del modelo de Holling-Tanner est√°n __garantizadas por el teorema de existencia y unicidad de Picard-Lindel√∂f__, siempre que las __funciones del lado derecho de las ecuaciones sean continuas y tengan derivadas parciales continuas con respecto a
$x$ y $y$__.

###Cantidades Conservadas o Acotadas

__Poblaci√≥n de Presas:__ La poblaci√≥n de presas $x(t)$ est√° acotada por la capacidad de carga
$K$, es decir, $0‚â§x(t)‚â§K$.

###Poblaci√≥n de Depredadores: La poblaci√≥n de depredadores
$y(t)$ est√° acotada por la disponibilidad de presas y la eficiencia de conversi√≥n, es decir, $0‚â§y(t)‚â§nx$.

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir s√≠mbolos y par√°metros
x, y = sym.symbols('x y')
r, K, q, a, s, n = 0.7, 10, 0.9, 0.75, 0.2, 0.7

# Definir las ecuaciones del modelo de Holling-Tanner
dx_dt = r * (1 - x / K - q / (x + a * y) * y) * x
dy_dt = s * (1 - y / (n * x)) * y

# Resolver las ecuaciones diferenciales ordinarias para encontrar los puntos cr√≠ticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Funci√≥n discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Funci√≥n para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Regi√≥n I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Regi√≥n II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Regi√≥n III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Regi√≥n IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Regi√≥n V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificaci√≥n"

# Evaluaci√≥n en puntos cr√≠ticos y clasificaci√≥n
print("Puntos cr√≠ticos y su clasificaci√≥n:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto cr√≠tico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificaci√≥n: {clas(J)}")

# Retrato fase del modelo de Holling-Tanner
def holling_tanner_model(X, t, r, K, q, a, s, n):
    x, y = X
    dx_dt = r * (1 - x / K - q / (x + a * y) * y) * x
    dy_dt = s * (1 - y / (n * x)) * y
    return [dx_dt, dy_dt]

# Definir el rango de valores para las gr√°ficas
x_range = np.linspace(-3, 22, 100)
y_range = np.linspace(-3, 10, 100)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = r * (1 - X / K - q / (X + a * Y) * Y) * X
dY_dt = s * (1 - Y / (n * X)) * Y

# Generar el retrato fase
fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(float(point[0]), float(point[1]), 'ro')
plt.xlabel('Poblaci√≥n de presas (x)')
plt.ylabel('Poblaci√≥n de depredadores (y)')
plt.title('Retrato Fase del Modelo de Holling-Tanner')
plt.xlim(-1, 22)
plt.ylim(-1, 10)
plt.grid()
plt.show()

# Graficar trayectorias del modelo de Holling-Tanner
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(0.1, 14, 5)
    for r0 in ic:
        for s0 in ic:
            x0 = [r0, s0]
            xs = odeint(func, x0, ts, args=(r, K, q, a, s, n))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Poblaci√≥n de presas (x)', fontsize=10)
    plt.ylabel('Poblaci√≥n de depredadores (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 15)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:15:10j, 0:10:10j]
    u = r * (1 - X / K - q / (X + a * Y) * Y) * X
    v = s * (1 - Y / (n * X)) * Y
    plt.quiver(X, Y, u, v, color='b')
    for point in critical_points:
        plt.plot(float(point[0]), float(point[1]), 'ro')
    plt.show()

# Ejecutar la gr√°fica de trayectorias
grafica(holling_tanner_model)
```

##Interpretaci√≥n

__Direcci√≥n de las trayectorias__:

 Las flechas indican la direcci√≥n del cambio en el espacio fase. En las regiones donde hay __un exceso de presas, la poblaci√≥n de depredadores crece debido a la disponibilidad de alimento__. A medida que los depredadores aumentan, la poblaci√≥n de presas disminuye, lo que eventualmente lleva a una __disminuci√≥n en la poblaci√≥n de depredadores debido a la escasez de alimento__.

__Puntos cr√≠ticos en las trayectorias__:

 Los puntos cr√≠ticos aparecen como puntos donde las trayectorias se intersectan o donde convergen hacia un punto estacionario. Estos puntos representan los __estados de equilibrio del sistema__, donde las poblaciones de presas y depredadores permanecen constantes en ausencia de perturbaciones externas.
