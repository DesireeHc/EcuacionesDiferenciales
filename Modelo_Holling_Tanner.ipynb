#Modelo de Holling Tanner



La relación entre depredadores y presas es un _fenómeno natural de la dinámica poblacional_, rama de la ecología, llamado _depredación_. En este tipo de fenómenos existen escenarios  tales como el __efecto indirecto de un depredador sobre la presa__, en este, basta la presencia del depredador, aún sin haber depredación, para que el __comportamiento y densidad de población  de la presa se vean afectados__. Este tipo de escenarios se pueden modelar con un sistema depredador-presa donde se busca representar dichos efectos indirectos.



El modelo fue desarrollado por __Crawford Stanley Holling__ y __John T. Tanner__. Holling, un ecólogo canadiense, es conocido por su trabajo en la respuesta funcional de los depredadores. _Introdujo la idea de diferentes tipos de respuestas funcionales_, siendo la Tipo II una de las más estudiadas. John T. Tanner, un matemático y ecólogo, contribuyó al desarrollo del modelo matemático en 1970. Su trabajo ayudó a _formalizar la dinámica poblacional de depredadores y presas usando ecuaciones diferenciales_. Tanner se centró en la aplicación de estos modelos para _comprender la estabilidad y las fluctuaciones cíclicas en las poblaciones naturales_.

Este modelo ha sido utilizado para estudiar sistemas _depredador-presa_, reproduciendo patrones observados en sistemas poblacionales, tales como _ciclos poblacionales_, _oscilaciones caóticas_ y _extinciones locales_.

Los modelos depredador-presa se clasifican en tres tipos, __los que dependen de la presa, los dependientes del depredador, o los mixtos__. Estas clasificaciones dependen de la rapidez con la que los depredadores eliminan a sus presas. Hollig propone tres tipos de respuestas funcionales, de los cuales, en su mayoria son aplicados a modelos que dependen de la presa.





__Tipo I.__

Esta respuesta funcional se obtiene suponiendo que __el
cambio en la densidad de poblacion de depredadores es proporcional a la densidad de la poblacion de presas disponibles__, es decir, existe un
_aumento lineal de la tasa de ataque de la poblacion de depredadores respecto a la densidad de poblacion de presas_, hasta llegar a un punto a
partir del cual la tasa maxima de ataque se hace
constante.

##¿Qué?

El modelo de Holling-Tanner es un __sistema de ecuaciones diferenciales__ que describe la __dinámica de una población de presas y una población de depredadores__. Este modelo tiene en cuenta una respuesta funcional tipo II, que es una representación más realista de la tasa de depredación en comparación con otros modelos, como el modelo Lotka-Volterra.




##¿Para qué?

El modelo se utiliza para entender __cómo interactúan las poblaciones de presas y depredadores en un ecosistema__. Ayuda a prever la _estabilidad de las poblaciones_ y las condiciones bajo las cuales las poblaciones pueden coexistir de manera sostenible. Este conocimiento es crucial en la gestión de la vida silvestre, la conservación de especies y el manejo de recursos naturales.





##¿Quién?
El modelo es de interés para ecólogos, biólogos, conservacionistas y gestores de recursos naturales. Investigadores académicos y estudiantes de biología matemática también utilizan este modelo para estudiar la dinámica de poblaciones.



##Ecuaciones

Estudiaremos la respuesta funcional Holling tipo II, el cual
tiene la siguiente forma:

$$h(x)=\frac{qx}{x+a}$$

donde $q>0$ y $a>0$
y $q$ es la tasa máxima de crecimiento del depredador y $a$ es la medida de protección a la presa brindada por el medio ambiente.

Sin embargo, muchos biólogos han cuestionado que este modelo solo dependa de la presa, por lo que se propone una respuesta funcional mixta, donde se dependa tanto de la presa como del depredador, a esto se le llama _razón- dependiente_, así se utilizará:

$$h(x)=\frac{qx}{x+ay}$$

En cualquier tiempo t, la __población de las presas__ se denota como $x$, mientras que la __población de los depredadores__ se denotarán como $y$. En este caso en particular, la población de la presa es mucho mayor que la del depredador, este modelo está dado por el siguiente sistema de ecuaciones:

$$\begin{cases}
\frac{dx}{dt}=r(1-\frac{x}{K}-\frac{q}{x+ay}y)x \\
\frac{dy}{dt}=s(1-\frac{y}{nx})y
\end{cases}
$$

##Parámetros


* $x$ : Población de presas.

* $y$ : Población de depredadores.


* $r$ : Tasa de crecimiento intrínseco para la presa (sin restricciones).

 * $0<r≤1$.
 Este parámetro describe la _velocidad a la cual la población de presas crece_ en ausencia de depredadores.


* $s$ : Tasa de crecimiento intrínseco para el depredador


* $K$ : Capacidad de carga del entorno para la población de presas.

 * $K>0$.
Este parámetro representa el _tamaño máximo de la población de presas que el ambiente puede soportar_.

* $a$ : Tasa máxima de depredación.

 * $a>0$.
Este parámetro indica la tasa a la cual los _depredadores consumen presas cuando las presas son abundantes_.

* $n$ : Medida de calidad de la presa como alimento para el depredador.

 * $q$ y $n$ deben ser __menores que uno__ para reflejar que __no todas las presas consumidas se convierten directamente en nuevos depredadores__; parte de la energía se pierde en el proceso.


##Interpretación de los Términos en las Ecuaciones



###Ecuación de la Población de Presas

$$\frac{dx}{dt}=r(1-\frac{x}{K}-\frac{q}{x+ay}y)x$$​



El término $1-\frac{x}{K}$ representa el __crecimiento logístico de las presas__, limitado por la __capacidad de carga__ $k$. El término $\frac{q}{x+ay}y$ representa la __tasa de depredación __ajustada por la __disponibilidad de presas__ y la __competencia entre depredadores__, el denominador, $x+ay$   representa la __dificultad de capturar presas__ a medida que aumentan las interacciones presa-depredador.

###Ecuación de la Población de Depredadores

$$\frac{dy}{dt}=s(1-\frac{y}{nx})$$


El término $1-\frac{y}{nx}$
  representa el __crecimiento de la población de depredadores__, limitado por la disponibilidad de presas. Si la cantidad de presas es insuficiente (
$y≫nx$), la población de depredadores disminuye. Si hay suficientes presas (
$$y≪nx$$), la población de depredadores puede crecer.

###Dinámica del Sistema


__Presas ($x$):__

* Crecen logísticamente con una capacidad de carga
$K$.

* Son depredadas por los depredadores $y$, con una _tasa de depredación que sigue una respuesta funcional tipo II de Holling_, donde la tasa de depredación se estabiliza a altos niveles de presas debido a la saturación del depredador.


__Depredadores ($y$):__

* Crecen en función de la disponibilidad de presas $x$.
* La tasa de crecimiento de los depredadores disminuye si la población de presas es insuficiente para sostener su crecimiento.

### Puntos criticos

Los puntos críticos representan los estados estacionarios del sistema, donde las __tasas de cambio de las poblaciones de presas y depredadores son cero__. Es decir, son las __soluciones donde las poblaciones de presas y depredadores no cambian con el tiempo__.

* __Punto crítico (0, 0)__: Ambas poblaciones están extintas. No hay presas ni depredadores presentes en el sistema.

* __Punto crítico donde
$𝑥≠0$ y $y≠0$__: Este punto crítico se determina resolviendo el sistema de ecuaciones igualado a cero.


__Por qué son importantes los puntos críticos:__

Los puntos críticos son fundamentales para entender las dinámicas del sistema predador-presa, la estabilidad de estos puntos críticos nos ayuda a prever __cómo responderá el sistema ante cambios en los parámetros del modelo o en las condiciones ambientales__.



##Teoremas Accesorios

###Existencia y Unicidad
La existencia y unicidad de soluciones para el sistema de ecuaciones diferenciales del modelo de Holling-Tanner están __garantizadas por el teorema de existencia y unicidad de Picard-Lindelöf__, siempre que las __funciones del lado derecho de las ecuaciones sean continuas y tengan derivadas parciales continuas con respecto a
$x$ y $y$__.

###Cantidades Conservadas o Acotadas

__Población de Presas:__ La población de presas $x(t)$ está acotada por la capacidad de carga
$K$, es decir, $0≤x(t)≤K$.

###Población de Depredadores: La población de depredadores
$y(t)$ está acotada por la disponibilidad de presas y la eficiencia de conversión, es decir, $0≤y(t)≤nx$.

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir símbolos y parámetros
x, y = sym.symbols('x y')
r, K, q, a, s, n = 0.7, 10, 0.9, 0.75, 0.2, 0.7

# Definir las ecuaciones del modelo de Holling-Tanner
dx_dt = r * (1 - x / K - q / (x + a * y) * y) * x
dy_dt = s * (1 - y / (n * x)) * y

# Resolver las ecuaciones diferenciales ordinarias para encontrar los puntos críticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")

# Retrato fase del modelo de Holling-Tanner
def holling_tanner_model(X, t, r, K, q, a, s, n):
    x, y = X
    dx_dt = r * (1 - x / K - q / (x + a * y) * y) * x
    dy_dt = s * (1 - y / (n * x)) * y
    return [dx_dt, dy_dt]

# Definir el rango de valores para las gráficas
x_range = np.linspace(-3, 22, 100)
y_range = np.linspace(-3, 10, 100)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = r * (1 - X / K - q / (X + a * Y) * Y) * X
dY_dt = s * (1 - Y / (n * X)) * Y

# Generar el retrato fase
fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(float(point[0]), float(point[1]), 'ro')
plt.xlabel('Población de presas (x)')
plt.ylabel('Población de depredadores (y)')
plt.title('Retrato Fase del Modelo de Holling-Tanner')
plt.xlim(-1, 22)
plt.ylim(-1, 10)
plt.grid()
plt.show()

# Graficar trayectorias del modelo de Holling-Tanner
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(0.1, 14, 5)
    for r0 in ic:
        for s0 in ic:
            x0 = [r0, s0]
            xs = odeint(func, x0, ts, args=(r, K, q, a, s, n))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Población de presas (x)', fontsize=10)
    plt.ylabel('Población de depredadores (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 15)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:15:10j, 0:10:10j]
    u = r * (1 - X / K - q / (X + a * Y) * Y) * X
    v = s * (1 - Y / (n * X)) * Y
    plt.quiver(X, Y, u, v, color='b')
    for point in critical_points:
        plt.plot(float(point[0]), float(point[1]), 'ro')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(holling_tanner_model)
```

##Interpretación

__Dirección de las trayectorias__:

 Las flechas indican la dirección del cambio en el espacio fase. En las regiones donde hay __un exceso de presas, la población de depredadores crece debido a la disponibilidad de alimento__. A medida que los depredadores aumentan, la población de presas disminuye, lo que eventualmente lleva a una __disminución en la población de depredadores debido a la escasez de alimento__.

__Puntos críticos en las trayectorias__:

 Los puntos críticos aparecen como puntos donde las trayectorias se intersectan o donde convergen hacia un punto estacionario. Estos puntos representan los __estados de equilibrio del sistema__, donde las poblaciones de presas y depredadores permanecen constantes en ausencia de perturbaciones externas.
