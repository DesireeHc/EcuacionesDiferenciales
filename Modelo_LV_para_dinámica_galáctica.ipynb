#Modelo de Lotka Volterra para din√°micas gal√°cticas

## __Datos hist√≥ricos, te√≥ricos y contextuales__

###__¬øQu√©?__

El modelo de Lotka-Volterra es un sistema de ecuaciones diferenciales no lineales de primer orden que describe la din√°mica de sistemas biol√≥gicos, como ya lo vimos con anterioridad, este es el modelo adaptado a la din√°mica de galaxias, el modelo puede describir __la interacci√≥n gravitacional entre dos galaxias__, como el gas interestelar y las estrellas, o entre distintas poblaciones estelares.

<img src="https://upload.wikimedia.org/wikipedia/commons/2/2f/Hubble_ultra_deep_field.jpg" alt="Texto Alternativo" width="270" height="230">


###__¬øPara qu√©?__

Este modelo se utiliza para __estudiar y predecir__ la evoluci√≥n de sistemas din√°micos complejos, en este caso, la interacci√≥n y evoluci√≥n de galaxias bajo fuerzas gravitacionales. Permite __comprender patrones y comportamientos emergentes en el tiempo__, como la formaci√≥n de estructuras, colisiones y fusiones de galaxias.




###__¬øQui√©n?__

El modelo original fue propuesto independientemente por Alfred J. Lotka en 1925 y Vito Volterra en 1926. En el contexto de la din√°mica de galaxias, investigadores y astrof√≠sicos han adaptado este modelo para estudiar las __interacciones gal√°cticas, como aquellas observadas en grupos de galaxias o durante colisiones gal√°cticas__.



### Contexto Hist√≥rico



__D√©cada de 1970 y 1980:__ Durante estas d√©cadas, la astrof√≠sica y la cosmolog√≠a empezaron a _adoptar m√°s formalmente m√©todos y modelos matem√°ticos para describir sistemas complejos_. La idea de usar modelos ecol√≥gicos como el de Lotka-Volterra se volvi√≥ m√°s prominente debido a la b√∫squeda de analog√≠as y modelos simplificados que pudieran describir procesos en sistemas astron√≥micos.



###Aplicaci√≥n en Din√°mica Gal√°ctica:
El modelo de Lotka-Volterra se adapt√≥ para estudiar la _evoluci√≥n de las galaxias_, particularmente en c√≥mo el gas interestelar se convierte en estrellas y c√≥mo las estrellas pueden influir en el gas interestelar. Esta adaptaci√≥n se basa en la idea de que las estrellas y el gas pueden interactuar de manera similar a como lo hacen los depredadores y las presas en los ecosistemas.



##__Ecuaciones del Modelo__

Las ecuaciones del modelo Lotka-Volterra adaptadas a la din√°mica de galaxias pueden formularse de manera similar al modelo original, con ciertas adaptaciones para tener en cuenta las interacciones gravitacionales.


__Componentes del Modelo:__

* __Poblaci√≥n de Gas Interestelar (G):__
 Representa la __cantidad de gas disponible en la galaxia__, que puede ser convertido en estrellas a trav√©s del proceso de formaci√≥n estelar.



* __Poblaci√≥n de Estrellas (S):__ Representa la __cantidad de estrellas en la galaxia__, que puede incrementar a medida que el gas se transforma en nuevas estrellas y decrecer por procesos como la muerte estelar y supernovas.

## Ecuaciones Diferenciales:

Las ecuaciones de Lotka-Volterra para estas dos poblaciones pueden formularse de la siguiente manera:

$$
\begin{cases}
\frac{dG}{dt}=\alpha G - Œ≤GS \\
\frac{dS}{dt}= Œ¥GS - Œ≥S
\end{cases}
$$

__donde:__

* $\alpha$ es la tasa de suministro de gas interestelar (puede ser positivo si hay suministro constante o cero si no hay nuevo gas).

*  $\beta$ es la tasa a la cual el gas se convierte en estrellas. Este par√°metro describe la _eficiencia del proceso de formaci√≥n estelar_. tipicamente var√° entre 0.01 y 1.

*  $\delta$ es la eficiencia de la formaci√≥n estelar a partir del gas.Puede ser comparable a $\beta$ y var√≠a entre 0.01 y 0.1.

* $\gamma$ es la tasa de disminuci√≥n de la poblaci√≥n estelar debido a la muerte estelar y otros procesos. Dependiendo del tipo de galaxia y su historia de formaci√≥n estelar, los valores de
Œ≥ pueden variar, pero t√≠picamente est√°n en el rango de 0.01 a 0.1.


## Interpretaci√≥n de las Ecuaciones

$$\frac{dG}{dt}=\alpha G - Œ≤GS$$

La primera ecuaci√≥n indica que el cambio en la cantidad de gas ($G$) depende de una fuente de suministro $\alpha G$ y una p√©rdida debido a la formaci√≥n de estrellas $\beta G S$.


$$\frac{dS}{dt}= Œ¥GS - Œ≥S$$

La segunda ecuaci√≥n indica que el cambio en la cantidad de estrellas ($S$) aumenta con la formaci√≥n de estrellas $\delta G S$ y disminuye con la muerte estelar $\gamma S$.



##Puntos de Equilibrio

Para encontrar los puntos de equilibrio, se establecen $\frac{dG}{dt} = 0$ y $\frac{dS}{dt} = 0$, y se resuelven las ecuaciones resultantes.

\begin{align*}
    G(\alpha - \beta S) &= 0 \\
    S(\delta G - \gamma) &= 0
\end{align*}

Los puntos cr√≠ticos son:

$(G,S)=(0,0)$

Este punto representa un __estado en el que no hay gas ni estrellas__. Es un estado trivial y generalmente no es de inter√©s en la din√°mica gal√°ctica real.

$(G,S)=(\frac{Œ≥}{Œ¥}, \frac{Œ±}{Œ≤})$

Este punto representa un __equilibrio donde las cantidades de gas y estrellas son constantes__. La relaci√≥n entre la tasa de suministro de gas y la tasa de formaci√≥n estelar, as√≠ como la tasa de disminuci√≥n de la poblaci√≥n estelar, determina las cantidades equilibradas.


###Teoremas

__Teorema (Existencia y Unicidad):__

La existencia y unicidad de soluciones para sistemas de ecuaciones diferenciales ordinarias se pueden garantizar utilizando el _Teorema de Picard-Lindel√∂f_. Este teorema establece que __si las funciones en el sistema son localmente Lipschitz continuas, entonces existe una √∫nica soluci√≥n que pasa por un punto inicial dado__.

__Teorema (Cantidad Conservada):__

Existe una cantidad conservada a lo largo de las trayectorias del sistema. Esta cantidad est√° dada por:

$$
H(G, S) = \alpha \ln S - \beta S + \gamma \ln G - \delta G
$$

__Teorema (Comportamiento Acotado):__

 Las soluciones del sistema de ecuaciones diferenciales del modelo de Lotka-Volterra est√°n acotadas y permanecen en el primer cuadrante del plano $(G,S)$ para
t‚â•0, siempre que las condiciones iniciales $(ùê∫_0,ùëÜ_0)$ est√©n en el primer cuadrante.

```python
# Importamos las librer√≠as necesarias
import sympy as sym  # Para c√°lculos simb√≥licos
import matplotlib.pyplot as plt  # Para gr√°ficos
import numpy as np  # Para c√°lculos num√©ricos
from scipy.integrate import odeint  # Para integrar ecuaciones diferenciales ordinarias

# Definir s√≠mbolos y par√°metros
G, S = sym.symbols('G S')
alpha, beta, delta, gamma = 0.5, 0.1, 0.3, 0.2  # Valores de ejemplo para los par√°metros

# Definir las ecuaciones del modelo de Lotka-Volterra para din√°mica gal√°ctica
dG_dt = alpha * G - beta * G * S
dS_dt = delta * G * S - gamma * S

# Encontrar los puntos cr√≠ticos
critical_points = sym.solve([dG_dt, dS_dt], (G, S))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dG_dt.diff(var) for var in (G, S)],
                            [dS_dt.diff(var) for var in (G, S)]])

# Funci√≥n discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Funci√≥n para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Regi√≥n I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Regi√≥n II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Regi√≥n III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Regi√≥n IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Regi√≥n V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificaci√≥n"

# Evaluaci√≥n en puntos cr√≠ticos y clasificaci√≥n
print("Puntos cr√≠ticos y su clasificaci√≥n:")
for point in critical_points:
    J = jacobi_matrix.subs({G: point[0], S: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto cr√≠tico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificaci√≥n: {clas(J)}")

# Retrato fase del modelo
def lotka_volterra_galactica(X, t, alpha, beta, delta, gamma):
    G, S = X
    dG_dt = alpha * G - beta * G * S
    dS_dt = delta * G * S - gamma * S
    return [dG_dt, dS_dt]

G_range = np.linspace(0, 10, 20)
S_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(G_range, S_range)
dG_dt_vals = np.array([[dG_dt.subs({G: G_val, S: S_val}) for G_val in G_range] for S_val in S_range], dtype=float)
dS_dt_vals = np.array([[dS_dt.subs({G: G_val, S: S_val}) for G_val in G_range] for S_val in S_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dG_dt_vals, dS_dt_vals, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Gas Interestelar (G)')
plt.ylabel('Estrellas (S)')
plt.title('Retrato Fase del Modelo de Lotka-Volterra en Din√°mica Gal√°ctica')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Lotka-Volterra
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 4)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(func, x0, ts, args=(alpha, beta, delta, gamma))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Gas Interestelar (G)', fontsize=10)
    plt.ylabel('Estrellas (S)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 10)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:10:10j, 0:10:10j]
    u = alpha * X - beta * X * Y
    v = delta * X * Y - gamma * Y
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gr√°fica de trayectorias
grafica(lotka_volterra_galactica)
```

##Trayectorias y Soluciones en el Retrato Fase

Las trayectorias en el retrato fase representan la evoluci√≥n temporal del sistema en el espacio de fase
(G,S). Cada trayectoria muestra c√≥mo cambian las cantidades de gas y estrellas con el tiempo, partiendo de diferentes condiciones iniciales.

## Retrato Fase
El retrato fase es una representaci√≥n gr√°fica de todas las posibles trayectorias del sistema en el espacio de fase. Ayuda a visualizar la din√°mica del sistema y la estabilidad de los puntos cr√≠ticos.

##Interpretaci√≥n de las Trayectorias

### Trayectorias Convergentes a (ùõæùõø,ùõºùõΩ):

__Interpretaci√≥n:__

 Si las trayectorias convergen a este punto cr√≠tico, indica que el sistema tiende a un estado de equilibrio estable donde las cantidades de gas y estrellas son constantes.

__Estabilidad:__ Un punto de equilibrio estable donde las trayectorias convergen es atractivo.

### Trayectorias Divergentes de
(0,0)

__Interpretaci√≥n:__

 Si las trayectorias divergen de
(0,0), indica que cualquier peque√±a cantidad de gas o estrellas hace que el sistema evolucione lejos del estado sin gas ni estrellas.

__Estabilidad:__ Un punto de equilibrio inestable donde las trayectorias divergen es repulsivo.

###Trayectorias C√≠clicas o Oscilatorias


__Interpretaci√≥n:__ Si las trayectorias son c√≠clicas, el sistema experimenta oscilaciones peri√≥dicas en las cantidades de gas y estrellas. Este comportamiento puede indicar interacciones complejas entre la formaci√≥n estelar y la disminuci√≥n estelar.

__Estabilidad:__ Las oscilaciones peri√≥dicas pueden indicar un equilibrio marginalmente estable o un ciclo l√≠mite.

## Aplicaciones y Limitaciones

* __Formaci√≥n y Evoluci√≥n Estelar:__

El modelo puede ser √∫til para estudiar la tasa de formaci√≥n estelar en diferentes tipos de galaxias y c√≥mo esta tasa se relaciona con la cantidad de gas disponible.

* __Interacci√≥n entre Componentes Gal√°cticos:__

Ayuda a entender c√≥mo las estrellas y el gas interestelar interact√∫an y afectan la evoluci√≥n de la galaxia.

* __Simplificaciones y Limitaciones:__

El modelo de Lotka-Volterra es una simplificaci√≥n y no captura todos los procesos complejos y feedbacks presentes en las galaxias, como la retroalimentaci√≥n de supernovas, la influencia de la materia oscura y la interacci√≥n con otras galaxias.


El uso del modelo Lotka-Volterra en din√°mica gal√°ctica proporciona una perspectiva matem√°tica simple pero poderosa para __analizar las interacciones b√°sicas entre diferentes componentes de una galaxia__. Aunque es una aproximaci√≥n simplificada, puede ofrecer valiosas intuiciones sobre los procesos fundamentales que rigen la evoluci√≥n de las galaxias.
