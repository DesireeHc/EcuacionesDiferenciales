4.1
b)
$$\frac{dx_1}{dt} = 2x_1+x_2$$
$$\frac{dx_2}{dt}= 3x_1+2x_2$$


```python
x, y=sym.symbols('x, y')
```

```python
f=sym.Eq(2*x+y, 0)
g=sym.Eq(3*x+2*y, 0)
```

```python

#me va a resolver el sistema (f,g) en (x, y)
sym.solve([f, g],(x,y))
```

```python
A=sym.Matrix([[2, 1],[3,2]])

#calcula el determinante
A.det()

A.trace()

#los eigenvalues separados
r_1, r_2=A.eigenvals()
A.eigenvects()
def D(M):
    return M.trace()**2-4*M.det()
```

```python
print("Det(A)=", A.det(), "Tr(A)=", A.trace(), "D(A)=", D(A))
```

```python
# The 2-dimensional linear system.
a, b, c, d = 2, 1, 3, 2
def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
    A= dx_dt
    print(dx_dt)
```

```python
ts = np.linspace(0, -4, 100)
ic = np.linspace(-1, 1, 5)
for r in ic:
    for s in ic:
        x0 = [r, s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[:,0], xs[:,1], "r-")# Label the axes and set fontsizes.

plt.xlabel('x', fontsize=15)  #coloca el nombre y tama√±o del cuadrante x
plt.ylabel('y', fontsize=15) #coloca el nombre y tama√±o del cuadrante y
plt.tick_params(labelsize=10) #tama√±o de par√°metros
plt.xlim(-1, 1)
plt.ylim(-1, 1)
X,Y = np.mgrid[-1:1:16j, -1:1:16j] #de donde a donde van la flechas, y cuantas flechas son
u = a*X + b*Y
v = c*X + d*Y
pl.quiver(X, Y, u, v, color = 'b')
plt.show()
```
