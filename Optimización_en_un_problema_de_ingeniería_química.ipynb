# Optimización en un problema de ingeniería química

Este problema de optimización se enmarca en el contexto de los procesos de fabricación, como en la ingeniería química, donde se busca maximizar o minimizar ciertos objetivos relacionados con el rendimiento del proceso.

\

Consideremos un proceso de fabricación que convierte determinadas materias primas en productos más valiosos. En este tipo de procesos de ingeniería química, suele haber una serie de respuestas, como:

 * Tasa de producción:
 Cuántos productos se generan en un periodo de tiempo.

* Rendimiento: La eficiencia con la que se transforma la materia prima en el producto deseado.

* Pureza de los distintos productos: la calidad o nivel de pureza del producto final.

Las cuales denotaremos como   $n_1, n_2,...,n_n$, se ven influidas por el nivel de variables controlables como **la temperatura, la presión, la concentración**, etc. A estas variables las llamaremos $x_1, x_2,..., x_m$.
Cada una de las respuestas $n_i​$ (como la tasa de producción o el rendimiento) dependerá de las variables controlables $x_1,x_2,..., x_m$​. Esto significa que las respuestas pueden ser expresadas como funciones matemáticas de las variables, es decir:

$$\eta_i=f_i(x_1,x_2,..., x_m)$$ con $i=1,..., n$

Aquí, $f_i$​ representa la relación entre las variables controlables y la respuesta $n_i​$. Estas funciones podrían ser **lineales o no lineales, dependiendo del tipo de proceso** y pueden no ser conocidas.


##El objetivo del problema de optimización:

El objetivo del ingeniero químico es **operar el proceso de fabricación** ( ajustar las variables $x_1, x_2,..., x_m$​) de la manera más eficiente posible u óptimo, y el quid de la cuestión reside en la forma en que definimos la palabra "eficientemente". "Óptimo" puede significar:

* Maximizar la tasa de producción,
* Maximizar el rendimiento,
* Aumentar la pureza del producto,

o una combinación de estos factores.


La dificultad es que no podemos decir simplemente que deseamos minimizar una sola cantidad, como el costo de fabricación de una unidad de producto, ya que intervienen otros factores. Por lo tanto, generalmente **deseamos mantener la pureza al mejor nivel posible y el proceso no debe demorar demasiado**.

\\

También puede ser necesario **restringir las variables directamente**; por ejemplo, ciertas presiones y temperaturas pueden ser inalcanzables o peligrosas.


Se puede argumentar que **debería ser posible obtener una función única de las respuestas**, por ejemplo, un criterio de rentabilidad, que mida por sí solo la conveniencia de un determinado conjunto de variables. En la práctica, no siempre es fácil obtener una función de este tipo. En el caso de la fabricación de un fármaco, por ejemplo, ¿qué valor monetario podría asignarse a las impurezas que podrían hacer que el fármaco fuera peligroso? ...

\\

En la práctica, el objetivo del ingeniero químico suele expresarse mejor en términos de *optimizar alguna respuesta principal*, generalmente relacionada con el costo, sujeta a ciertas restricciones tanto en las otras respuestas **(las variables dependientes)** como en las variables controlables **(las variables independientes)**. La "respuesta principal" corresponde aquí a la "función de costo" que se optimizó en el problema anteriormente visto (localización óptima). Por supuesto, debe enfatizarse que se necesita una **consideración cuidadosa**, en cada caso, para decidir qué cantidad debe optimizarse, porque si nos engañamos y optimizamos el criterio equivocado, nuestros esfuerzos **pueden conducir a pérdidas en lugar de ahorros**.

La situación se ilustra en la figura siguiente, donde la respuesta principal es el costo del producto por unidad de peso $n_1$, y una respuesta auxiliar es la impureza del producto en alguna unidad apropiada.

Los contornos del costo $n$ y la impureza $n_2$ son mostradas en el espacio de las variables $x₁$ presión y $x_a$ temperatura. Si el **objetivo es minimizar el costo 1 sujeto a la restricción de que la presión no debe ser mayor que un valor $x₁ = X$ y el nivel de impurezas no mayor que $0.2$**, el problema es encontrar ese punto de costo mínimo dentro del área restringida denotada por el sombreado en la Figura Este es de hecho el punto $P$.

##Función objetivo:

En términos matemáticos, el problema general puede plantearse de la siguiente manera: deseamos **minimizar una respuesta principal**. Se debe definir una función que represente lo que se desea maximizar o minimizar. Por ejemplo, si el objetivo es maximizar la tasa de producción $n_1$​, la función objetivo sería:

$$\eta_1=f_i(x_1,x_2,..., x_m)$$



sujeto a restricciones en las respuestas auxiliares de la forma

$$\alpha_i \leq \eta_i=f_i(x_1,x_2,..., x_m) \leq \beta_i$$


y a las restricciones sobre las variables independientes

$$\alpha_k \leq g_k \leq \beta_k$$ con $k=1, 2, ..., n$

donde las $g_k$ son funciones dadas de $xy$ (funciones de restricción) y $\beta_k$​ son los límites permitidos.

\\

En esta forma, el problema matemático es obviamente mucho más general que cualquiera de los problemas considerados en el problema visto con anterioridad. En él se nos dieron las formas de las funciones $f$, mientras que aquí no se conocen de antemano. No había restricciones presentes, aunque podrían haberse introducido simplemente diciendo que ciertas áreas no estaban disponibles para la ubicación de plantas de energía o máquinas.

Solución del problema:
Para resolver este problema, se pueden utilizar métodos de optimización como:
Programación lineal (si las funciones f_i​ y g_j​ son lineales),
Métodos de optimización no lineal (si las funciones son no lineales),
Métodos numéricos, como el gradiente descendente, que ajustan las variables x1,x2,..., xm​ para encontrar el punto que maximiza o minimiza la función objetivo.

Para resolver con un enfoque simplificado, podemos utilizar la idea de aproximación local de las funciones que describen el problema, como las tasas de producción o el rendimiento de un proceso. Para hacerlo manejable, se utiliza la serie de Taylor y suposiciones sobre la suavidad de las funciones.


## 1. Aproximación local lineal:

La clave es que, **en lugar de trabajar con las funciones exactas**, que pueden ser complejas, se hace una **aproximación de las funciones en un punto cercano al que nos interesa estudiar**. Este enfoque se conoce como aproximación local y en este caso, la función se aproxima mediante un plano (es decir, una función lineal).

Si tenemos una función $f(x_1, x_2,..., x_m)$ que depende de varias variables $x_1, x_2,..., x_m$, y suponemos que f es lo suficientemente suave (diferenciable), podemos aproximarla localmente alrededor de un punto $(x_1^0, x_2^0, ..., x_m^0)$ utilizando su primera derivada. Esto nos da una **aproximación lineal de la función en ese punto**, que es esencialmente el primer término de su serie de Taylor:

$$f(x1​,x2​,...,xm​)≈f(x_1^0, x_2^0, ..., x_m^0)+i=1∑m​∂xi​∂f(x_1^0, x_2^0, ..., x_m^0)(xi​−xi0​)$$


Esta expresión representa la función como **un plano tangente al punto $(x_1^0, x_2^0, ..., x_m^0)$**, y nos permite trabajar con una versión más simple de la función que puede ser optimizada más fácilmente.




### Ejemplo de Optimización por Aproximación Local Lineal

##Problema

Supongamos que queremos maximizar la tasa de producción de un reactor químico que depende de la *temperatura* ($T$) y la *presión* ($P$). La tasa de producción $R(T, P)$ alrededor de un punto de operación

$(T_0, P_0) = (300 \, **K, 5 \, *atm*)$

es **aproximada localmente mediante una función lineal**. Además, tenemos una restricción de que la suma de temperatura y presión no debe exceder un valor dado.

Queremos aproximar la tasa de producción $R(T, P)$ en términos lineales alrededor de $(T_0, P_0)$ y maximizarla, respetando la restricción.

##Paso 1: Expansión lineal de $R(T, P)$

Sabemos que $R(T, P)$ puede ser aproximada por un plano tangente en el punto $(T_0, P_0)$ usando su derivada en dicho punto. La aproximación local lineal de $R(T, P)$ es:

$
R(T, P) \approx R(T_0, P_0) + \frac{\partial R}{\partial T} (T - T_0) + \frac{\partial R}{\partial P} (P - P_0)
$

Supongamos que se han calculado las siguientes derivadas parciales en el punto $(T_0, P_0) = (300 \, \text{K}, 5 \, \text{atm})$:

$
\frac{\partial R}{\partial T} = 0.5 \, \text{mol/h·K}, \quad \frac{\partial R}{\partial P} = 1 \, \text{mol/h·atm}
$

Y supongamos que la tasa de producción en el punto $R(300 \, \text{K}, 5 \, \text{atm}) = 100 \, \text{mol/h}$.

Por lo tanto, la aproximación lineal sería:

$
R(T, P) \approx 100 + 0.5 (T - 300) + 1 (P - 5)
$

##Paso 2: Restricciones

Se tiene una **restricción operativa**: la suma de la temperatura y la presión no debe exceder cierto valor, por ejemplo:

$
T + P \leq 310
$

Esta restricción limita las combinaciones de $T$ y $P$ que podemos utilizar para maximizar la tasa de producción.

##Paso 3: Plantear el problema de optimización

El problema es maximizar $R(T, P)$, sujeto a la restricción $T + P \leq 310$. Esto puede formularse como:

$
\text{Maximizar } R(T, P) = 100 + 0.5 (T - 300) + 1 (P - 5)
$

\\

Sujeto a la restricción:

$T + P \leq 310$

##Paso 4: Resolución

Queremos maximizar $R(T, P)$ y al mismo tiempo cumplir con $T + P \leq 310$. Vamos a utilizar la restricción $T + P = 310$ (es decir, alcanzamos el límite de la restricción, ya que estamos maximizando).

Podemos expresar $T$ en términos de $P$ usando la restricción:

$
T = 310 - P
$

Sustituyendo esto en la función objetivo:

$
R(T, P) = 100 + 0.5 (310 - P - 300) + 1 (P - 5)
$

Simplificamos:

$
R(T, P) = 100 + 0.5 (10 - P) + 1 (P - 5)
$
$
R(T, P) = 100 + 5 - 0.5P + P - 5
$
$R(T, P) = 100 + 0.5P$

##Paso 5: Maximización

Para maximizar $R(T, P) = 100 + 0.5P$, vemos que el valor de $P$ debe ser lo más grande posible bajo la restricción. Dado que $T + P = 310$ y $T \geq 0$, el valor máximo de $P$ ocurre cuando $T = 0$, es decir, $P = 310$.

Sustituyendo $P = 310$ en la función de tasa de producción:

$R(T = 0, P = 310) = 100 + 0.5(310) = 100 + 155 = 255 \, \text{mol/h}$

##Paso 6: Interpretación del resultado}

La tasa de producción óptima es $255 \, \text{mol/h}$, que se alcanza cuando la presión es $P = 310 \, \text{atm}$ y la temperatura es $T = 0 \, \text{K}$. Aunque este valor de temperatura no es físicamente realista, el resultado sirve para ilustrar la aproximación y el proceso de optimización.



```python
import sympy as sp

# Definimos las variables
x1, x2 = sp.symbols('x1 x2')

# Función original
f = 3*x1 + 2*x2

# Punto alrededor del cual hacemos la aproximación lineal
x10 = 1
x20 = 1

# Gradiente de la función
grad_f = [sp.diff(f, var) for var in (x1, x2)]

# Aproximación lineal de la función en (x10, x20)
approx_linear = f.subs({x1: x10, x2: x20})
approx_linear += sum([grad * (var - val) for grad, var, val in zip(grad_f, (x1, x2), (x10, x20))])

print("Aproximación lineal:")
sp.pprint(approx_linear)
```

## 2. Expansión cuadrática:



Posteriormente, el texto menciona que se podría ir un paso más allá, considerando no solo términos lineales, sino también términos cuadráticos. Esto implica utilizar más términos de la serie de Taylor, lo que da una mejor aproximación de la función si ésta tiene curvatura (por ejemplo, si es una parábola).


La expansión cuadrática de una función alrededor de un punto $(x_1^0, x_2^0, ..., x_m^0)$ sería:

$f(x_1, x_2,..., x_m)≈f(x_1^0, x_2^0, ..., x_m^0)+∑_{i=1}^{m} \frac{∂f}{∂x_i}(x_1^0, x_2^0, ..., x_m^0)(x_i - x_i^0) + \frac{1}{2} \sum_{i=1}^{m} \sum_{j=1}^{m} \frac{\partial^2 f}{\partial x_i \partial x_j} (x_1^0, x_2^0, ..., x_m^0) (x_i - x_i^0)(x_j - x_j^0)$

En este caso, se incluye una matriz de segundas derivadas (la matriz Hessiana) que *captura la curvatura de la función*. Esto es especialmente útil en problemas donde la función objetivo no es lineal, y queremos una mejor aproximación de su comportamiento cerca de un punto crítico.


##Ejemplo de Optimización con Expansión Cuadrática


###Problema

Imagina que estás optimizando la tasa de producción $R(T, P)$ de un proceso químico que depende de la *temperatura* $T$ y la *presión* $P$. Se sabe que $R(T, P)$ tiene un comportamiento no lineal, por lo que utilizamos una aproximación cuadrática en torno al punto de operación $(T_0, P_0) = (300 \, \text{K}, 5 \, \text{atm})$.

\\

###Paso 1: Expansión Cuadrática

La expansión cuadrática de $R(T, P)$ alrededor del punto $(T_0, P_0)$ es:

$R(T, P) \approx R(T_0, P_0) + \frac{\partial R}{\partial T}(T_0, P_0) (T - T_0) + \frac{\partial R}{\partial P}(T_0, P_0) (P - P_0) + \frac{1}{2} \left[ \frac{\partial^2 R}{\partial T^2}(T_0, P_0) (T - T_0)^2 + 2 \frac{\partial^2 R}{\partial T \partial P}(T_0, P_0) (T - T_0)(P - P_0) + \frac{\partial^2 R}{\partial P^2}(T_0, P_0) (P - P_0)^2 \right]$

Vamos a suponer que:

* $R(T_0, P_0) = 100 \, \text{mol/h}$,

* $\frac{\partial R}{\partial T}(T_0, P_0) = 0.5 \, \text{mol/h·K}$,

* $\frac{\partial R}{\partial P}(T_0, P_0) = 1 \, \text{mol/h·atm}$,

* $\frac{\partial^2 R}{\partial T^2}(T_0, P_0) = -0.01 \, \text{mol/h·K}^2$,

* $\frac{\partial^2 R}{\partial P^2}(T_0, P_0) = -0.02 \, \text{mol/h·atm}^2$,

* $\frac{\partial^2 R}{\partial T \partial P}(T_0, P_0) = 0.01 \, \text{mol/h·K·atm}$.

Sustituyendo estos valores en la expansión cuadrática, obtenemos:

$R(T, P) \approx 100 + 0.5 (T - 300) + 1 (P - 5) - \frac{0.01}{2} (T - 300)^2 - \frac{0.02}{2} (P - 5)^2 + 0.01 (T - 300)(P - 5)$

\\

##Paso 2: Plantear el problema de optimización

Queremos maximizar $R(T, P)$ usando la expansión cuadrática y, además, tenemos una restricción de que la suma de temperatura y presión no debe exceder $310$, es decir:

$T + P \leq 310$

\\

##Paso 3: Resolución

Utilizamos la restricción $T + P = 310$ para simplificar el problema. Expresamos $T$ en términos de $P$:

$T = 310 - P$

Sustituyendo esto en la expresión de $R(T, P)$:

$R(310 - P, P) \approx 100 + 0.5 (310 - P - 300) + 1 (P - 5) - 0.005 (310 - P - 300)^2 - 0.01 (P - 5)^2 + 0.01 (310 - P - 300)(P - 5)$

Simplificamos:

$R(310 - P, P) \approx 100 + 0.5(10 - P) + 1(P - 5) - 0.005(10 - P)^2 - 0.01(P - 5)^2 + 0.01(10 - P)(P - 5)$

\\

##Paso 4: Derivada para maximización

Para encontrar el valor óptimo de $P$, derivamos $R(T, P)$ con respecto a $P$ y encontramos el valor de $P$ que maximiza $R$:

$\frac{dR}{dP} = 0.5(-1) + 1 - 0.01(2)(10 - P)(-1) - 0.02(P - 5) + 0.01((10 - P) + (P - 5))$

Simplificando la derivada:

$\frac{dR}{dP} = -0.5 + 1 + 0.02(10 - P) - 0.02(P - 5) + 0.01(10 - P + P - 5)$


Igualamos la derivada a cero para encontrar el punto crítico:

$-0.5 + 1 + 0.02(10 - P) - 0.02(P - 5) + 0.05 = 0$

Resolviendo la ecuación:

$0.55 + 0.02(15 - 2P) = 0$

$0.55 + 0.3 - 0.04P = 0$

$0.85 - 0.04P = 0$

$P = \frac{0.85}{0.04} = 21.25$

\\

##Paso 5: Solución óptima

El valor óptimo de $P$ es $P = 21.25 \, \text{atm}$. Usando la restricción $T + P = 310$, obtenemos el valor óptimo de $T$:

$T = 310 - 21.25 = 288.75 \, \text{K}$

\\

##Paso 6: Interpretación del resultado

La tasa de producción óptima se alcanza cuando $T = 288.75 \, \text{K}$ y $P = 21.25 \, \text{atm}$. Sustituyendo estos valores en la expresión cuadrática para $R(T, P)$, se puede calcular el valor máximo de la tasa de producción.


```python
import sympy as sp

# Definimos las variables
T, P = sp.symbols('T P')

# Función original
R = 100 + 0.5*(T - 300) + 1*(P - 5) - 0.005*(T - 300)**2 - 0.01*(P - 5)**2 + 0.01*(T - 300)*(P - 5)

# Gradiente de R
grad_R = [sp.diff(R, var) for var in (T, P)]

# Matriz Hessiana (segunda derivada de R)
hessian_R = [[sp.diff(grad, var) for var in (T, P)] for grad in grad_R]

# Expansión cuadrática en torno a (T0, P0) = (300, 5)
T0, P0 = 300, 5
R0 = R.subs({T: T0, P: P0})

# Aproximación cuadrática
approx_quad = R0 + sum([grad*(var - val) for grad, var, val in zip(grad_R, (T, P), (T0, P0))])
approx_quad += 0.5 * sum([hessian_R[i][j]*(var1 - val1)*(var2 - val2)
                          for i, var1, val1 in zip(range(2), (T, P), (T0, P0))
                          for j, var2, val2 in zip(range(2), (T, P), (T0, P0))])

print("Expansión cuadrática:")
sp.pprint(approx_quad)
```

## 3. Resolución manual:

Para un ejemplo sencillo que se pueda resolver a mano, podemos concentrarnos en el caso lineal:
Supongamos que queremos maximizar una función

$ f(x1,x2)=3x1+2x2f(x_1, x_2) = 3x_1 + 2x_2f(x1​,x2​)=3x1​+2x2$​, que podría representar, por ejemplo, la producción en función de la temperatura $x_1$ y la presión $x_2$​.

\\

Si hay restricciones, como $x1+x2≤5x_1 + x_2 \leq 5x1​+x2​≤5$ (por ejemplo, una limitación en los recursos disponibles), podemos resolver este problema a mano usando métodos como multiplicadores de Lagrange o el método gráfico si el número de variables es pequeño.

Este enfoque aproximado y paso a paso permite tratar **problemas de optimización con funciones suaves** utilizando herramientas que simplifican su análisis, facilitando así la resolución a mano o con métodos más sencillos.


Resolveremos un ejemplo mediante el método gráfico.

##Optimización utilizando el método gráfico


###Problema

Supongamos que queremos maximizar la función objetivo:

$
Z = 3x_1 + 2x_2
$

sujeta a las siguientes restricciones:

\begin{align*}
x_1 + x_2 &\leq 4 \\
2x_1 + x_2 &\leq 6 \\
x_1, x_2 &\geq 0
\end{align*}



Definición del problema: El objetivo es maximizar la función $Z = 3x_1 + 2x_2$ sujeta a tres restricciones.

##Paso 1: Representación gráfica de las restricciones

Para resolver este problema gráficamente, trazamos las restricciones en un plano cartesiano. Vamos a representar las restricciones como ecuaciones de rectas:


* $x_1 + x_2 = 4$: Intersecciones con los ejes $x_1$ y $x_2$ son $(4, 0)$ y $(0, 4)$.

* $2x_1 + x_2 = 6$: Intersecciones con los ejes $x_1$ y $x_2$ son $(3, 0)$ y $(0, 6)$.


Graficamos estas rectas y la región factible definida por las restricciones.

Se incluyen las ecuaciones de las restricciones y sus intersecciones con los ejes.

##Paso 2: Evaluación en los vértices

Los vértices de la región factible son los puntos $(0, 0)$, $(3, 0)$, $(2, 2)$ y $(0, 4)$. Evaluamos la función objetivo en estos puntos:

* $Z(0, 0) = 3(0) + 2(0) = 0$
* $Z(3, 0) = 3(3) + 2(0) = 9$
* $Z(2, 2) = 3(2) + 2(2) = 10$
* $Z(0, 4) = 3(0) + 2(4) = 8$

Se calculan los valores de la función objetivo en los vértices de la región factible.
Solución óptima: Se determina que el punto $(2, 2)$ maximiza la función objetivo, dando un valor de $Z_{\text{máx}} = 10$.

##Paso 3: Solución óptima

El valor máximo de $Z$ se alcanza en el punto $(2, 2)$, donde $Z = 10$. Por lo tanto, la solución óptima es:

$
x_1 = 2, \quad x_2 = 2, \quad Z_{\text{máx}} = 10
$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones de las restricciones
def restriccion1(x1):
    return 4 - x1  # x1 + x2 <= 4

def restriccion2(x1):
    return 6 - 2*x1  # 2x1 + x2 <= 6

# Generamos los puntos en el eje x1
x1_vals = np.linspace(0, 5, 100)

# Graficamos las restricciones
plt.plot(x1_vals, restriccion1(x1_vals), label=r'$x_1 + x_2 = 4$', color='blue')
plt.plot(x1_vals, restriccion2(x1_vals), label=r'$2x_1 + x_2 = 6$', color='red')

# Sombreado de la región factible
plt.fill_between(x1_vals, np.minimum(restriccion1(x1_vals), restriccion2(x1_vals)), 0, where=(x1_vals <= 3), color='gray', alpha=0.5)

# Etiquetas y límites
plt.xlim((0, 5))
plt.ylim((0, 5))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.legend()

# Puntos de intersección
plt.plot([0, 3, 2, 0], [0, 0, 2, 4], 'ro')

# Solución óptima
plt.text(2, 2, 'Óptimo (2,2)', horizontalalignment='right')

plt.show()
```

A primera vista, puede parecer sorprendente que podamos decir algo sobre el problema muy general. La clave del tratamiento es que investigamos sólo el **comportamiento local de las funciones**.
