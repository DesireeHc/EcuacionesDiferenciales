La pendiente de la curva de oferta es positiva, lo que indica que, __a medida que el precio aumenta, los productores est√°n dispuestos a ofrecer m√°s cantidad del bien o servicio.__

##Modelo

Sea $p = P(t)$ la __funci√≥n precio de un bien__ en el tiempo y $Q(t)$ es la __cantidad del bien__ en el tiempo $ùë°$. El n√∫mero de
unidades del bien que _desean los consumidores_ por unidad de tiempo, en
cualquier tiempo $t$ se llama __demanda__ y se denota por $D = D(t)$. Esta demanda adem√°s de depender del precio $p$ en cualquier tiempo $t$, tambi√©n depende de la direcci√≥n en la cual los consumidores creen que tomar√°n los precios, esto es, la __tasa de cambio del precio__ $P‚Äô(t)$. En matem√°ticas, la dependencia de $D(t)$, $P(t)$, y de $P‚Äô(t)$ se puede escribir como:

$$D=f(p(t),p(t))$$

Por tanto, $S$ es la funci√≥n de oferta.


Para que las anteriores consideraciones tengan sentido, se debe asumir
lo siguiente:

__a) Econom√≠a competitiva y libre.__

 Es decir, que los consumidores y
productores compiten para determinar los precios.

__b) No hay demora en el suministro.__

 En la ecuaci√≥n se asume que los
productores usan una tasa de cambio del precio en el tiempo, esto para
decidir sobre la oferta disponible. Esto es una aproximaci√≥n a
la realidad, puesto que en la pr√°ctica hay una __demora entre el tiempo
de producci√≥n real y el mercadeo al consumidor__. En tal caso se reemplazar√≠a por:

$$S=(g(p(t-T), p'(t-T))$$

Sin embargo, se tomar√° con la suposici√≥n inicial.

__c) No se consideran los precios de otros bienes.__

 Los precios de otros bienes en el mercado no se toman en cuenta.

__d) Los precios, demanda y oferta son continuos.__

 Los precios toman valores discretos, pero en la pr√°ctica, se pueden __aproximar__ con un buen grado de precisi√≥n adoptando __valores continuos.__

##Ecuaciones del Modelo

Las ecuaciones diferenciales representan un __modelo cl√°sico de oferta y demanda__ donde $x$ representa la oferta y $y$ la demanda. Las ecuaciones describen c√≥mo la oferta y la demanda cambian en funci√≥n del tiempo y de las interacciones entre ellas.

El modelo cl√°sico de oferta y demanda se puede representar por las siguientes ecuaciones diferenciales:

$$
\begin{cases}
\frac{dx}{dt}=Œ±x‚àíŒ≤xy \\
\frac{dy}{dt}=‚àíŒ≥y+Œ¥xy
\end{cases}
$$

Donde:
* x representa la (oferta).
* y representa la (demanda).
* Œ±,Œ≤,Œ≥,Œ¥ son __par√°metros positivos__ que representan tasas de crecimiento y tasas de interacci√≥n entre las poblaciones.

### Interpretaci√≥n de las ecuaciones

1.  La tasa de cambio de la oferta con respecto al tiempo $\frac{dx}{dt}$ se expresa como:

 $$\frac{dx}{dt} = \alpha x - \beta xy$$

Donde:
* $\alpha$ representa la tasa de crecimiento natural de la oferta (en ausencia de demanda).
* $\beta$ mide c√≥mo la demanda afecta la oferta (por ejemplo, si la demanda es alta, la oferta puede disminuir).


2. La tasa de cambio de la demanda con respecto al tiempo $\frac{dy}{dt}$ se expresa como:

$$\frac{dy}{dt} = -\gamma y + \delta xy$$

Donde:

* $\gamma$ representa la tasa de decrecimiento natural de la demanda.
* $\delta$ mide c√≥mo la oferta afecta la demanda (por ejemplo, si la oferta es alta, la demanda puede disminuir, c√≥mo  la cantidad demandada aumenta debido a factores como el crecimiento de la poblaci√≥n, cambios en el gusto de los consumidores, entre otros).

Estas ecuaciones capturan la din√°mica entre la oferta y la demanda en un mercado. El equilibrio se alcanza cuando las tasas de cambio de ambas son iguales, lo que determina el precio y la cantidad de equilibrio.
Esto da lugar a los puntos cr√≠ticos:

### Puntos Cr√≠ticos y su Clasificaci√≥n
__Puntos Cr√≠ticos:__

Los puntos cr√≠ticos son aquellos donde las tasas de cambio de $x$ y $y$ son cero simult√°neamente. En este modelo, los puntos cr√≠ticos se calculan resolviendo:


$$Œ±x‚àíŒ≤xy=0$$
$$‚àíŒ≥y+Œ¥xy=0$$

As√≠, los puntos cr√≠ticos para este modelo son:

* $(0,0)$

Esto indica que no hay demanda ni oferta del bien en el mercado.

* $(0,\frac{Œ≥}{Œ¥})$

Esto significa que no hay demanda del bien, pero hay una oferta espec√≠fica en el mercado.

* $(\frac{Œ±}{Œ≤},0)$

Esto indica que hay demanda espec√≠fica del bien, pero no hay oferta correspondiente en el mercado.

* $(\frac{ùõæ}{ùõø},\frac{ùõº}{ùõΩ})$

Esto indica que hay un equilibrio en el mercado donde la cantidad demandada y la cantidad ofrecida son iguales.

## Teoremas

Este modelo cl√°sico de oferta y demanda en econom√≠a se fundamenta en varios principios y teoremas importantes que sustentan su an√°lisis y predicciones:


* __Ley de la demanda:__ Esta ley establece que,  manteniendo todo lo dem√°s constante, __la cantidad demandada de un bien disminuye cuando su precio aumenta, y viceversa__.

<img src="https://angelacastaneda.net/wp-content/uploads/2023/06/La-Demanda_Angela-Castaneda-1-1024x576.jpg" alt="Texto Alternativo" width="300" height="200">




* __Ley de la oferta:__ Esta ley indica que, la __cantidad ofrecida de un bien aumenta cuando su precio aumenta, y disminuye cuando su precio disminuye__. Es el principio b√°sico que describe la relaci√≥n positiva entre el precio de un bien y la cantidad que los productores est√°n dispuestos a ofrecer.

<img src="https://angelacastaneda.net/wp-content/uploads/2023/06/La-Oferta_Angela-Castaneda-1-1024x576.jpg" alt="Texto Alternativo" width="300" height="200">




* __Equilibrio de mercado:__ Este teorema establece que en un mercado competitivo y libre de restricciones, __el precio y la cantidad de un bien se establecen en el punto donde la cantidad demandada es igual a la cantidad ofrecida__. Es decir, donde las curvas de oferta y demanda se intersectan.





* __Teorema de la mano invisible de Adam Smith:__  Sugiere que los individuos que buscan maximizar su propio beneficio econ√≥mico (en un mercado competitivo) de manera indirecta __promueven el bienestar general de la sociedad__ a trav√©s del mecanismo de precios y el intercambio voluntario.



* __Teorema de la eficiencia de Pareto:__ Este teorema indica que en un mercado competitivo, el equilibrio de mercado es eficiente en el sentido de Pareto, lo que significa que __no es posible mejorar la situaci√≥n de un individuo sin empeorar la de otro__. En otras palabras, el equilibrio maximiza el excedente total del consumidor y el productor.




Estos teoremas y principios proporcionan __la base te√≥rica y conceptual__ para entender c√≥mo funcionan los mercados en el modelo de oferta y demanda

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definir s√≠mbolos y par√°metros
x, y = sym.symbols('x y')
alpha, beta, gamma, delta = 1, 0.5, 1.5, 0.8

# Definir las ecuaciones del modelo de oferta y demanda
dx_dt = alpha * x - beta * x * y
dy_dt = -gamma * y + delta * x * y

# Resolver las ecuaciones diferenciales ordinarias para encontrar los puntos cr√≠ticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Funci√≥n discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Funci√≥n para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Regi√≥n I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Regi√≥n II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Regi√≥n III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Regi√≥n IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Regi√≥n V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificaci√≥n"

# Evaluaci√≥n en puntos cr√≠ticos y clasificaci√≥n
print("Puntos cr√≠ticos y su clasificaci√≥n:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto cr√≠tico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificaci√≥n: {clas(J)}")

# Retrato fase del modelo de oferta y demanda
def oferta_demanda_model(X, t, alpha, beta, gamma, delta):
    x, y = X
    dx_dt = alpha * x - beta * x * y
    dy_dt = -gamma * y + delta * x * y
    return [dx_dt, dy_dt]

# Definir el rango de valores para las gr√°ficas
x_range = np.linspace(0, 12, 100)
y_range = np.linspace(0, 12, 100)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = alpha * X - beta * X * Y
dY_dt = -gamma * Y + delta * X * Y

# Generar el retrato fase
fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(float(point[0]), float(point[1]), 'ro')
plt.xlabel('Oferta(x)')
plt.ylabel('Demanda (y)')
plt.title('Retrato Fase del Modelo de Oferta y Demanda')
plt.xlim(0, 12)
plt.ylim(0, 12)
plt.grid()
plt.show()

# Graficar trayectorias del modelo de oferta y demanda
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(0.1, 10, 5)
    for x0 in ic:
        for y0 in ic:
            initial_conditions = [x0, y0]
            xs = odeint(func, initial_conditions, ts, args=(alpha, beta, gamma, delta))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Oferta (x)', fontsize=10)
    plt.ylabel('Demanda (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 12)
    plt.ylim(0, 12)

    X, Y = np.mgrid[0:12:10j, 0:12:10j]
    u = alpha * X - beta * X * Y
    v = -gamma * Y + delta * X * Y
    plt.quiver(X, Y, u, v, color='b')
    for point in critical_points:
        plt.plot(float(point[0]), float(point[1]), 'ro')
    plt.show()

# Ejecutar la gr√°fica de trayectorias
grafica(oferta_demanda_model)
```

## Interpretaci√≥n del Retrato Fase

__Puntos Cr√≠ticos:__

 Los puntos cr√≠ticos representan los puntos de equilibrio del sistema, donde la oferta y la demanda se igualan.

__Direcciones de las Trayectorias:__

 Las flechas en el retrato fase indican la direcci√≥n en la que evolucionan las variables $ùë•$ y $ùë¶$ a partir de condiciones iniciales dadas. Por ejemplo, si comenzamos con un __exceso de oferta__ (por encima del punto de equilibrio), las trayectorias nos muestran c√≥mo la oferta disminuir√° y la demanda aumentar√° hasta alcanzar nuevamente el equilibrio.

__Estabilidad de los Puntos Cr√≠ticos:__

 La estabilidad de los puntos cr√≠ticos se puede analizar observando c√≥mo se comportan las trayectorias alrededor de estos puntos. Por ejemplo, un punto cr√≠tico es __estable__ si las trayectorias cercanas convergen hacia √©l con el tiempo, mientras que es __inestable__ si las trayectorias se alejan de √©l.

## Conclusiones del Modelo de Oferta y Demanda


###Ajuste Autom√°tico:

El modelo sugiere que __los mercados tienden hacia el equilibrio de manera autom√°tica a trav√©s de los ajustes en los precios y las cantidades ofrecidas y demandadas__. Este proceso es fundamentalmente impulsado por las decisiones de los consumidores y los productores en respuesta a cambios en las condiciones econ√≥micas.

###Impacto de los Cambios en los Par√°metros:

Variaciones en los par√°metros del modelo (tales como cambios en los costos de producci√≥n, impuestos, preferencias de los consumidores, etc.) pueden desplazar las curvas de oferta y demanda, afectando as√≠ el precio y la cantidad de equilibrio. El retrato fase puede mostrar c√≥mo estos cambios __afectan las trayectorias de oferta y demanda en el tiempo__.

###Pol√≠tica Econ√≥mica:

 El modelo puede utilizarse para evaluar el __impacto de pol√≠ticas econ√≥micas__, como subsidios, impuestos o regulaciones, sobre el equilibrio del mercado. Por ejemplo, un subsidio a la oferta podr√≠a desplazar la curva de oferta hacia la derecha, aumentando la cantidad ofrecida y reduciendo el precio de equilibrio.

### An√°lisis de Sensibilidad:
 El an√°lisis del retrato fase tambi√©n permite evaluar la __sensibilidad del sistema a perturbaciones externas o cambios abruptos en las condiciones econ√≥micas__. Por ejemplo, c√≥mo reaccionar√≠a el mercado a una escasez repentina de materias primas o un cambio repentino en la demanda debido a factores externos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the system of differential equations
def model(z, t, alpha, beta, gamma, delta):
    x, y = z
    dxdt = alpha * x - beta * x * y
    dydt = -gamma * y + delta * x * y
    return [dxdt, dydt]

# Parameters
alpha = 0.9
beta = 0.5
gamma = 0
delta = 10.8

# Initial conditions
z0 = [1.0, 1.0]

# Time vector
t = np.linspace(0, 10, 100)

# Solve the system of differential equations
z = odeint(model, z0, t, args=(alpha, beta, gamma, delta))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(z[:, 0], z[:, 1], label='Trajectory')
plt.plot(0, 0, 'ro', label='(0, 0)')
plt.plot(0, gamma/delta, 'go', label='(0, Œ≥/Œ¥)')
plt.plot(alpha/beta, 0, 'bo', label='(Œ±/Œ≤, 0)')
plt.plot(gamma/delta, alpha/beta, 'ko', label='(Œ≥/Œ¥, Œ±/Œ≤)')
plt.xlabel('Cantidad ofrecida, y')
plt.ylabel('Cantidad demandada, x')
plt.title('Diagrama de fase del modelo de oferta y demanda')
plt.legend()
plt.grid(True)
plt.show()
```
