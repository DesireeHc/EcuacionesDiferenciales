#Modelo de Belousov-Zhabotinsky (BZ)



Las reacciones químicas consisten en _sustancias químicas llamadas reactivos_, que en contacto, reaccionan entre sí, dando un _producto_. Después de un teimpo se __alcanza un equilibrio__, donde la relación entre reactivos y productos permanecen _constantes_. Es decir, sólo queda una mezcla entre productos y reactivos.

En cuanto a las reacciones oscilantes, existe una __situación fuera del equilibrio__, es el momento antes de que pase el tiempo necesario para llegar al equilibrio. Si las oscilaciones son periódicas, se le llama _regular_, en caso contrario será un régimen _caótico_.

El modelo de Belousov-Zhabotinsky (BZ) es un modelo famoso para **reacciones químicas oscilatorias**. Para simplificar y adaptar a un sistema 2x2, usaremos una forma reducida del modelo de Oregonator (modelo teórico para un tipo de reacción autocatalítica), que es una simplificación común del BZ. En este modelo, __las variables de estado son las concentraciones de los intermedios de reacción__, es decir que estas concentraciones __representan  las cantidades que evolucionan con el tiempo y caracterizan el estado actual del sistema__.



##Introducción y Contexto Histórico




###¿Qué?

El modelo de Belousov-Zhabotinsky (BZ) describe una _reacción química oscilante_ que fue descubierta por Boris Belousov en la década de 1950 y más tarde investigada en detalle por Anatol Zhabotinsky. Es un ejemplo clásico de un _sistema lejos del equilibrio_ que exhibe comportamiento oscilante debido a la retroalimentación no lineal en las reacciones químicas.




###¿Para qué?

Este modelo se utiliza para estudiar __fenómenos de autoorganización y patrones espaciales en sistemas químicos__. También sirve como un ejemplo de cómo las __reacciones no lineales pueden llevar a oscilaciones y patrones complejos__ en lugar de alcanzar un estado de equilibrio simple. Puede considerarse como un ejemplo de la teoría del caos.



###¿Quién?

Este modelo fue propuesto por _Boris Belousov_ en 1951 y posteriormente desarrollado por _Anatol Zhabotinsky_, quien explica la secuencia de la reacción química. Es conocido por ser uno de los primeros ejemplos de un sistema químico que exhibe _autoorganización y autooscilaciones_.

## Ecuaciones del Modelo


El modelo BZ se puede describir mediante un __sistema de ecuaciones diferenciales ordinarias (EDOs)__ que representan las __concentraciones de los intermediarios químicos en función del tiempo__. Un modelo simplificado se basa en las siguientes ecuaciones:



$$\begin{cases}
\frac{dX}{dt} = A - (B + 1)X + X^2Y \\
\frac{dY}{dt} = BX - X^2Y
\end{cases}
$$


donde:
* $X$ es la concentración de un intermediario reactivo,
* $Y$ es la concentración de otro intermediario reactivo (p, ej. ácido),
* $A$ y $B$ son parámetros que representan las tasas de entrada y salida de los reactivos y las constantes de velocidad de las reacciones.


##Interpretación del Sistema
###Primera Ecuación:

$$\frac{dX}{dt} = A - (B + 1)X + X^2Y $$


* Esta ecuación describe la tasa de cambio de la concentración de $x$(bromuro).


* $A$ representa una _fuente constante de bromuro_.


* $(B+1)x$ representa la _pérdida de bromuro_ debido a reacciones que lo consumen.


* $x^2 y$ representa la _retroalimentación positiva_, donde la concentración de bromuro aumenta debido a la interacción con el intermediario $y$.



###Segunda Ecuación:



$$\frac{dY}{dt} = BX - X^2Y$$


* Esta ecuación describe la _tasa de cambio de la concentración_ de $y$ (intermediario).


* $Bx$ representa la _producción de
$y$_ a partir de la reacción con $x$.


* $x^2y$ representa la _pérdida de
$y$_ debido a su consumo en la reacción.


###Comportamiento Oscilante

El sistema muestra un comportamiento oscilante debido a la __dinámica de retroalimentación entre las concentraciones de $x$ y $y$__. El comportamiento oscilante se puede entender de la siguiente manera:

* __Retroalimentación Positiva:__

Cuando la concentración de
$x$ es alta, esto aumenta la tasa de producción de
$y$, lo que a su vez puede aumentar la concentración de $x$ debido a la retroalimentación positiva en
$x^2y$.

* __Retroalimentación Negativa:__

A medida que $x$ y $y$ aumentan, la tasa de reacción que consume
$y$ (es decir, $x^2y$) también aumenta, lo que eventualmente disminuye la concentración de
$y$. Esta disminución en
$y$ reduce la retroalimentación positiva, lo que lleva a una disminución en $x$, y el ciclo comienza de nuevo.

Para que el sistema muestre estas oscilaciones químicas, los parámetros deben estar dentro de los siguientes rangos:

* $A$ debe ser un valor __positivo relativamente pequeño__.

* $B$ debe ser un valor __positivo mayor que $A$ pero no excesivamente grande__.

###Puntos Críticos

Bajo las suposiciones dadas, se tiene el único punto crítico:
$$(x, y)=(a, \frac{b}{a})$$

sin considerar variaciones extremas

###Matriz Jacobiana

La matriz Jacobiana del sistema es:

$$
J = \begin{pmatrix}
\frac{\partial f}{\partial X} & \frac{\partial f}{\partial Y} \\
\frac{\partial g}{\partial X} & \frac{\partial g}{\partial Y}
\end{pmatrix}
= \begin{pmatrix}
-(B + 1) + 2XY & X^2 \\
B - 2XY & -X^2
\end{pmatrix}
$$


##Tipo de Punto Crítico

Para clasificar los puntos críticos, evaluamos la traza $(t)$ y el determinante $(d)$ de la matriz Jacobiana.

Estas ecuaciones describen cómo __las concentraciones de las especies químicas cambian con el tiempo debido a las reacciones químicas que tienen lugar__.

Dependiendo de los valores de los parámetros 𝐴 y 𝐵, el sistema puede exhibir __comportamientos oscilatorios__, como la formación de patrones espaciales y la autoorganización.

El modelo de Belousov-Zhabotinsky ha sido de gran interés en la investigación de la __autoorganización y la emergencia de patrones en sistemas químicos y biológicos__, y ha inspirado estudios en campos como la química teórica, la biofísica y la neurociencia.




```python
# Importamos las librerías necesarias
import sympy as sym  # Para cálculos simbólicos
import matplotlib.pyplot as plt  # Para gráficos
import numpy as np  # Para cálculos numéricos
from scipy.integrate import odeint  # Para integrar ecuaciones diferenciales ordinarias

# Definir símbolos y parámetros
x, y = sym.symbols('x y')
a, b = .5, 1.0  # Valores de ejemplo para los parámetros

# Definir las ecuaciones del modelo de Oregonator (simplificación del modelo BZ)
dx_dt = a - (b + 1) * x + x**2 * y
dy_dt = b * x - x**2 * y

# Resolver las ecuaciones diferenciales ordinarias para encontrar los puntos críticos
# Encontrar los puntos críticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")

# Retrato fase del modelo BZ
def bz_model(X, t, a, b):
    x, y = X
    dx_dt = a - (b + 1) * x + x**2 * y
    dy_dt = b * x - x**2 * y
    return [dx_dt, dy_dt]

# Definir el rango de valores para las gráficas
x_range = np.linspace(-1, 3, 20)
y_range = np.linspace(1, 5, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

# Generar el retrato fase
fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(float(point[0]), float(point[1]), 'ro')
plt.xlabel('Concentración de intermedio 1 (x)')
plt.ylabel('Concentración de intermedio 2 (y)')
plt.title('Retrato Fase del Modelo BZ')
plt.xlim(-1, 3)
plt.ylim(1, 5)
plt.grid()
plt.show()

# Graficar trayectorias del sistema BZ
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(0, 1.8, 4)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(func, x0, ts, args=(a, b))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Concentración de intermedio 1 (x)', fontsize=10)
    plt.ylabel('Concentración de intermedio 2 (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 2)
    plt.ylim(0, 2.2)

    X, Y = np.mgrid[0:2:10j, 0:2.5:10j]
    u = a - (b + 1) * X + X**2 * Y
    v = b * X - X**2 * Y
    plt.quiver(X, Y, u, v, color='b')
    for point in critical_points:
        plt.plot(float(point[0]), float(point[1]), 'ro')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(bz_model)
```

Este es un _modelo de laboratorio de muchos fenómenos bioquímicos que oscilan_, tales como el titilar de las luciernagas, los ritmos circadianos, o el ciclo de Krebs de la respiración celular.


###Análisis

__Autoorganización:__

 El retrato fase del modelo BZ puede mostrar regiones de espacio donde las concentraciones de las especies químicas _exhiben patrones espaciales y temporales autoorganizados_. Estos patrones pueden surgir debido a la interacción no lineal entre las diferentes especies químicas y sus reacciones.

__Estabilidad y Inestabilidad:__

 La _clasificación de los puntos críticos del sistema_ nos permite determinar si estos puntos son _estables o inestables_.
