## Ecuaciones del Modelo


El modelo BZ se puede describir mediante un __sistema de ecuaciones diferenciales ordinarias (EDOs)__ que representan las __concentraciones de los intermediarios qu铆micos en funci贸n del tiempo__. Un modelo simplificado se basa en las siguientes ecuaciones:



$$\begin{cases}
\frac{dX}{dt} = A - (B + 1)X + X^2Y \\
\frac{dY}{dt} = BX - X^2Y
\end{cases}
$$


donde:
* $X$ es la concentraci贸n de un intermediario reactivo,
* $Y$ es la concentraci贸n de otro intermediario reactivo (p, ej. 谩cido),
* $A$ y $B$ son par谩metros que representan las tasas de entrada y salida de los reactivos y las constantes de velocidad de las reacciones.


##Interpretaci贸n del Sistema
###Primera Ecuaci贸n:

$$\frac{dX}{dt} = A - (B + 1)X + X^2Y $$


* Esta ecuaci贸n describe la tasa de cambio de la concentraci贸n de $x$(bromuro).


* $A$ representa una _fuente constante de bromuro_.


* $(B+1)x$ representa la _p茅rdida de bromuro_ debido a reacciones que lo consumen.


* $x^2 y$ representa la _retroalimentaci贸n positiva_, donde la concentraci贸n de bromuro aumenta debido a la interacci贸n con el intermediario $y$.



###Segunda Ecuaci贸n:



$$\frac{dY}{dt} = BX - X^2Y$$


* Esta ecuaci贸n describe la _tasa de cambio de la concentraci贸n_ de $y$ (intermediario).


* $Bx$ representa la _producci贸n de
$y$_ a partir de la reacci贸n con $x$.


* $x^2y$ representa la _p茅rdida de
$y$_ debido a su consumo en la reacci贸n.


###Comportamiento Oscilante

El sistema muestra un comportamiento oscilante debido a la __din谩mica de retroalimentaci贸n entre las concentraciones de $x$ y $y$__. El comportamiento oscilante se puede entender de la siguiente manera:

* __Retroalimentaci贸n Positiva:__

Cuando la concentraci贸n de
$x$ es alta, esto aumenta la tasa de producci贸n de
$y$, lo que a su vez puede aumentar la concentraci贸n de $x$ debido a la retroalimentaci贸n positiva en
$x^2y$.

* __Retroalimentaci贸n Negativa:__

A medida que $x$ y $y$ aumentan, la tasa de reacci贸n que consume
$y$ (es decir, $x^2y$) tambi茅n aumenta, lo que eventualmente disminuye la concentraci贸n de
$y$. Esta disminuci贸n en
$y$ reduce la retroalimentaci贸n positiva, lo que lleva a una disminuci贸n en $x$, y el ciclo comienza de nuevo.

Para que el sistema muestre estas oscilaciones qu铆micas, los par谩metros deben estar dentro de los siguientes rangos:

* $A$ debe ser un valor __positivo relativamente peque帽o__.

* $B$ debe ser un valor __positivo mayor que $A$ pero no excesivamente grande__.

###Puntos Cr铆ticos

Bajo las suposiciones dadas, se tiene el 煤nico punto cr铆tico:
$$(x, y)=(a, \frac{b}{a})$$

sin considerar variaciones extremas

###Matriz Jacobiana

La matriz Jacobiana del sistema es:

$$
J = \begin{pmatrix}
\frac{\partial f}{\partial X} & \frac{\partial f}{\partial Y} \\
\frac{\partial g}{\partial X} & \frac{\partial g}{\partial Y}
\end{pmatrix}
= \begin{pmatrix}
-(B + 1) + 2XY & X^2 \\
B - 2XY & -X^2
\end{pmatrix}
$$


##Tipo de Punto Cr铆tico

Para clasificar los puntos cr铆ticos, evaluamos la traza $(t)$ y el determinante $(d)$ de la matriz Jacobiana.

Estas ecuaciones describen c贸mo __las concentraciones de las especies qu铆micas cambian con el tiempo debido a las reacciones qu铆micas que tienen lugar__.

Dependiendo de los valores de los par谩metros  y , el sistema puede exhibir __comportamientos oscilatorios__, como la formaci贸n de patrones espaciales y la autoorganizaci贸n.

El modelo de Belousov-Zhabotinsky ha sido de gran inter茅s en la investigaci贸n de la __autoorganizaci贸n y la emergencia de patrones en sistemas qu铆micos y biol贸gicos__, y ha inspirado estudios en campos como la qu铆mica te贸rica, la biof铆sica y la neurociencia.




```python
# Importamos las librer铆as necesarias
import sympy as sym  # Para c谩lculos simb贸licos
import matplotlib.pyplot as plt  # Para gr谩ficos
import numpy as np  # Para c谩lculos num茅ricos
from scipy.integrate import odeint  # Para integrar ecuaciones diferenciales ordinarias

# Definir s铆mbolos y par谩metros
x, y = sym.symbols('x y')
a, b = .5, 1.0  # Valores de ejemplo para los par谩metros

# Definir las ecuaciones del modelo de Oregonator (simplificaci贸n del modelo BZ)
dx_dt = a - (b + 1) * x + x**2 * y
dy_dt = b * x - x**2 * y

# Resolver las ecuaciones diferenciales ordinarias para encontrar los puntos cr铆ticos
# Encontrar los puntos cr铆ticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                            [dy_dt.diff(var) for var in (x, y)]])

# Funci贸n discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Funci贸n para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Regi贸n I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Regi贸n II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Regi贸n III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Regi贸n IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Regi贸n V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificaci贸n"

# Evaluaci贸n en puntos cr铆ticos y clasificaci贸n
print("Puntos cr铆ticos y su clasificaci贸n:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto cr铆tico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificaci贸n: {clas(J)}")

# Retrato fase del modelo BZ
def bz_model(X, t, a, b):
    x, y = X
    dx_dt = a - (b + 1) * x + x**2 * y
    dy_dt = b * x - x**2 * y
    return [dx_dt, dy_dt]

# Definir el rango de valores para las gr谩ficas
x_range = np.linspace(-1, 3, 20)
y_range = np.linspace(1, 5, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

# Generar el retrato fase
fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(float(point[0]), float(point[1]), 'ro')
plt.xlabel('Concentraci贸n de intermedio 1 (x)')
plt.ylabel('Concentraci贸n de intermedio 2 (y)')
plt.title('Retrato Fase del Modelo BZ')
plt.xlim(-1, 3)
plt.ylim(1, 5)
plt.grid()
plt.show()

# Graficar trayectorias del sistema BZ
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(0, 1.8, 4)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(func, x0, ts, args=(a, b))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Concentraci贸n de intermedio 1 (x)', fontsize=10)
    plt.ylabel('Concentraci贸n de intermedio 2 (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 2)
    plt.ylim(0, 2.2)

    X, Y = np.mgrid[0:2:10j, 0:2.5:10j]
    u = a - (b + 1) * X + X**2 * Y
    v = b * X - X**2 * Y
    plt.quiver(X, Y, u, v, color='b')
    for point in critical_points:
        plt.plot(float(point[0]), float(point[1]), 'ro')
    plt.show()

# Ejecutar la gr谩fica de trayectorias
grafica(bz_model)
```
