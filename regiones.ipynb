```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl
```

```python
#a, b, c, d=2, 1, 1, 1
# Define a 2x2 matrix
A = [[a, b], [c, d]]
a=A[0][0]
b=A[0][1]
c=A[1][0]
d=A[1][1]
# Access matrix elements
print(a)  # Output: 2 (Accessing element at row 0, column 0)
print(d)  # Output: 1 (Accessing element at row 1, column 1)
print(A)
```

```python
from ast import AnnAssign
def det(A):
  return[a*d-b*c]
def tr(A):
  return[a+d]
#def tr2(A): return [(A[0][0]*A[1][1]-A[0][1]*A[1][0])**2]
#def det4(A): return[4*(A[0][0]*A[1][1]-A[0][1]*A[1][0])]
#def D(A): return [tr2(A)-det4(A)] COMO PUEDO HACERLO MÁS PEQUEÑO
def D(A):
  return[(a+d)**2-4*(a*d-c*b)]
```

```python
#defina la matriz random
def generate_random_matrices(n, min_value, max_value):
    matrices = []
    for _ in range(n):
        matrix = np.random.randint(min_value, max_value + 1, size=(2, 2))
        matrices.append(matrix)
    return matrices
```

```python
#dar determinante y traza
def calculate_trace(matrix):
    trace = np.trace(matrix)
    return trace

def calculate_determinant(matrix):
    determinant = np.linalg.det(matrix)
    return determinant
```

```python
def calculate_discriminant(matrix):
    b = matrix[0][1]
    a = matrix[0][0]
    c = matrix[1][0]

    discriminant = (b ** 2) - (4 * a * c)
    return discriminant
```

```python
n_matrices = 1000  # Number of matrices to generate
min_value = -10  # Minimum value for random elements
max_value = 10  # Maximum value for random elements

matrices = generate_random_matrices(n_matrices, min_value, max_value)
```

```python
traces = []
determinants = []
discriminants = []

for matrix in matrices:
    trace = calculate_trace(matrix)
    determinant = calculate_determinant(matrix)
    discriminant = calculate_discriminant(matrix)

    traces.append(trace)
    determinants.append(determinant)
    discriminants.append(discriminant)
```

```python
plt.figure(figsize=(10, 6))

# Colormap for discriminant values
cmap = plt.cm.get_cmap('viridis')

# Scatter plot with color-coding based on discriminant
plt.scatter(traces, determinants, c=discriminants, cmap=cmap)

# Label axes
plt.xlabel('Trace')
plt.ylabel('Determinant')

# Add colorbar
plt.colorbar(label='Discriminant')

# Set title and show plot
plt.title('Trace-Determinant Plane with Discriminant Color-Coding')
plt.show()
```
