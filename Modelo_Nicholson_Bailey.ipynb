#Modelo Nicholson-Bailey:

##Datos históricos, teóricos y contextuales
###¿Qué es el Modelo Nicholson-Bailey?

El Modelo Nicholson-Bailey, también conocido como el Modelo de Competencia Interespecífica, es un modelo matemático que describe la __dinámica de dos poblaciones que compiten por los mismos recursos__. Fue desarrollado por los biólogos británicos Charles Nicholson y Maurice Bailey en 1935 para estudiar la interacción entre dos especies de escarabajos que habitaban el mismo almacén de grano.

###¿Para qué se utiliza el Modelo Nicholson-Bailey?

El Modelo Nicholson-Bailey se utiliza para comprender __la dinámica de las poblaciones que compiten por los mismos recursos en diversos ecosistemas__. Se ha aplicado a una amplia gama de sistemas biológicos, incluyendo la competencia entre plantas, animales y microorganismos. El modelo también se ha utilizado para estudiar la interacción entre depredadores y presas.

###¿Quién desarrolló el Modelo Nicholson-Bailey?

El Modelo Nicholson-Bailey fue desarrollado por los biólogos británicos Charles Nicholson y Maurice Bailey en 1935. Nicholson era un entomólogo y Bailey era un matemático. Ambos trabajaban en el Instituto de Investigación Agrícola Rothamsted en Harpenden, Inglaterra. A principios del siglo XX, la ecología experimentó un importante desarrollo como disciplina científica. Los científicos comenzaron a estudiar las __interacciones entre los organismos y su entorno, incluyendo la competencia entre especies por recursos limitados__.

Antes del Modelo Nicholson-Bailey, ya existían otros modelos que describían la competencia entre especies. Sin embargo, estos modelos eran más simples y no consideraban la capacidad de carga del medio ambiente ni los coeficientes de competencia.

## Análisis Matemático del Modelo Nicholson-Bailey para Parásito-Hospedador

El Modelo Nicholson-Bailey, originalmente desarrollado para describir la competencia entre dos especies, se ha adaptado para __modelar la dinámica de poblaciones de parásitos y hospedadores__. En este contexto, el modelo describe la _interacción entre una población de parásitos que se alimenta de un hospedador y la población de hospedadores que albergan a los parásitos_.

##Ecuaciones del Modelo:

Las ecuaciones del Modelo Nicholson-Bailey para parásito-hospedador son similares a las del modelo original, pero con algunas modificaciones para considerar la naturaleza biológica de la interacción parásito-hospedador. Las ecuaciones son:

Para la población de hospedadores $(H)$:

$$\frac{dH}{dt} = rH - \frac{aHP}{K}$$

Donde:

* $H$: Representa la densidad de la población de hospedadores en un momento dado.
* $r$: Representa la tasa de crecimiento natural de la población de hospedadores en ausencia de parásitos.
* $a$: Representa el coeficiente de competencia por contacto entre hospedadores, que refleja la eficiencia con la que los parásitos se transmiten entre hospedadores.
* $P$: Representa la densidad de la población de parásitos en un momento dado.
* $K$: Representa la capacidad de carga del medio ambiente para la población de hospedadores.
Para la población de parásitos $(P)$:

$$\frac{dP}{dt} = cβHP - μP$$

Donde:

* $P$: Representa la densidad de la población de parásitos en un momento dado.
* $c$: Representa la tasa de conversión de hospedadores infectados a nuevos parásitos.
* $β$: Representa la tasa de contacto entre parásitos y hospedadores susceptibles.
* $H$: Representa la densidad de la población de hospedadores en un momento dado.
* $μ$: Representa la tasa de mortalidad natural de los parásitos.

###Explicación de los términos:

* Tasa de crecimiento natural $(r)$: Representa la tasa a la que la población de hospedadores crecería en ausencia de parásitos.

* Coeficiente de competencia por contacto $(a)$: Representa la eficiencia con la que los parásitos se transmiten entre hospedadores.

* Tasa de conversión $(c)$: Representa la tasa a la que los hospedadores infectados se convierten en nuevos parásitos.

* Tasa de contacto $(β)$: Representa la tasa a la que los parásitos entran en contacto con hospedadores susceptibles.

* Tasa de mortalidad natural $(μ)$: Representa la tasa a la que los parásitos mueren naturalmente.

* Capacidad de carga $(K)$: Representa el tamaño máximo de la población de hospedadores que el medio ambiente puede sostener.
Interpretación de las ecuaciones:

La primera ecuación describe cómo __la tasa de crecimiento de la población de hospedadores se ve afectada por su propia tasa de crecimiento natural__, la competencia por contacto entre hospedadores y la presión de los parásitos.

La segunda ecuación describe cómo __la tasa de crecimiento de la población de parásitos se ve afectada por la tasa de conversión de hospedadores infectados__, la tasa de contacto entre parásitos y hospedadores, la densidad de la población de hospedadores y la tasa de mortalidad natural de los parásitos.

##Análisis del Modelo:

El análisis del Modelo Nicholson-Bailey para parásito-hospedador se realiza principalmente a través del estudio de su retrato fase. El retrato fase muestra las trayectorias de las poblaciones de hospedadores y parásitos en el plano (H, P) y permite identificar los equilibrios del sistema, su estabilidad y las condiciones que conducen a la erradicación del parásito, la persistencia del parásito y la coexistencia estable de parásitos y hospedadores.

##Ecuaciones del Modelo Nicholson-Bailey

Las ecuaciones del Modelo Nicholson-Bailey son un sistema de dos ecuaciones diferenciales de primer orden que describen las tasas de cambio de las poblaciones de las dos especies competidoras. Las ecuaciones son:

$$
\left\{
\begin{array}{ll}
\frac{dH}{dt} = rH - \frac{aHP}{K} \\
\frac{dP}{dt} = cβHP - μP
\end{array}
\right.
$$



##Puntos críticos

Los puntos críticos del Modelo Nicholson-Bailey son los puntos en el plano (X, Y) donde las tasas de cambio de ambas poblaciones son cero. Se pueden encontrar resolviendo el sistema de ecuaciones:

$$\frac{dH}{dt} = 0$$

$$\frac{dP}{dt} = 0$$

Las soluciones son:

* $(0, 0)$: Equilibrio de extinción, donde ambas poblaciones se extinguen.
* $(K, 0)$: Equilibrio de exclusión competitiva, donde la población de hospedadores persiste en su capacidad de carga (K) y la población de parásitos se extingue


La estabilidad de los puntos críticos se puede analizar utilizando la matriz jacobiana. La matriz jacobiana evaluada en cada punto crítico proporciona información sobre el comportamiento local del sistema alrededor de ese punto.


__Equilibrio de extinción (0, 0):__

 Si la traza de la matriz jacobiana en (0, 0) es positiva y su determinante es negativo, entonces el equilibrio de extinción es un punto silla inestable. Esto significa que pequeñas perturbaciones en las poblaciones cerca del equilibrio de extinción conducirán a que las poblaciones se alejen de ese punto.

__Equilibrio de exclusión competitiva (K, 0):__

 Si la traza de la matriz jacobiana en (K, 0) es negativa y su determinante es positivo, entonces el equilibrio de exclusión competitiva es un punto estable. Esto significa que pequeñas perturbaciones en las poblaciones cerca del equilibrio de exclusión competitiva conducirán a que las poblaciones regresen a ese punto.

##Retrato fase

El retrato fase del Modelo Nicholson-Bailey es una representación gráfica de las trayectorias de las poblaciones de las dos especies competidoras a lo largo del tiempo. __Las trayectorias pueden converger a un equilibrio, indicando que el conflicto se resuelve de manera estable, o pueden divergir, indicando que el conflicto continúa escalando__.

## Condiciones de Existencia y Unicidad:

###Teorema de Existencia:

Se puede demostrar que existe al menos una solución no trivial para las ecuaciones del Modelo Nicholson-Bailey para parásito-hospedador bajo ciertas condiciones sobre los parámetros del modelo. Esto significa que el modelo puede exhibir una dinámica no trivial con poblaciones de parásitos y hospedadores que no son constantes.

### Teorema de Unicidad:

Se puede demostrar que la solución del Modelo Nicholson-Bailey para parásito-hospedador es única bajo ciertas condiciones sobre los parámetros del modelo. Esto significa que para un conjunto dado de parámetros, solo existe un posible resultado para las poblaciones de parásitos y hospedadores a lo largo del tiempo.

2. Cantidades Conservadas o Acotadas:

###Cantidad Conservada:

En el Modelo Nicholson-Bailey para parásito-hospedador, la cantidad total de parásitos y hospedadores puede ser conservada bajo ciertas condiciones. Esto significa que la suma de las poblaciones de parásitos y hospedadores permanece constante en el tiempo.

###Cantidades Acotadas:
En el Modelo Nicholson-Bailey para parásito-hospedador, las poblaciones de parásitos y hospedadores pueden estar acotadas superiormente o inferiormente bajo ciertas condiciones. Esto significa que las poblaciones no pueden crecer o disminuir sin límite, sino que están restringidas a un rango específico de valores.

3. Estabilidad de Equilibrios:

Teorema de Estabilidad Local: Se puede analizar la estabilidad local de los equilibrios del Modelo Nicholson-Bailey para parásito-hospedador utilizando la matriz jacobiana. Si la matriz jacobiana evaluada en un equilibrio tiene autovalores con partes reales negativas, entonces el equilibrio es localmente estable. Esto significa que pequeñas perturbaciones en las poblaciones de parásitos y hospedadores cerca del equilibrio no causarán un cambio significativo en el comportamiento del sistema.

###Teorema de Estabilidad Global:
 En algunos casos, se puede demostrar la estabilidad global de los equilibrios del Modelo Nicholson-Bailey para parásito-hospedador. Esto significa que cualquier perturbación en las poblaciones de parásitos y hospedadores, sin importar su tamaño, eventualmente conducirá al sistema de regreso al equilibrio.


```python
# Importamos las librerias necesarias
import sympy as sym # Para calculos simbolicos
import matplotlib.pyplot as plt # Para graficos
import numpy as np # Para calculos numericos
from scipy.integrate import odeint # Para integrar ecuaciones diferenciales ordinarias
import pylab as pl # Para graficos adicionales

# Definir símbolos y parámetros
H, P = sym.symbols('H P')
r, K, a, c, d = 1.0, 100, 0.1, 0.1, 0.1

# Definir las ecuaciones del modelo de Nicholson-Bailey
dH_dt = r * H * (1 - H / K) - a * H * P
dP_dt = c * a * H * P - d * P

# Encontrar los puntos críticos
critical_points = sym.solve([dH_dt, dP_dt], (H, P))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dH_dt.diff(var) for var in (H, P)],
                           [dP_dt.diff(var) for var in (H, P)]])

# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()

# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"

# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({H: point[0], P: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")


# Retrato fase del modelo
def nicholson_bailey(X, t, r, K, a, c, d):
    H, P = X
    dH_dt = r * H * (1 - H / K) - a * H * P
    dP_dt = c * a * H * P - d * P
    return [dH_dt, dP_dt]

x_range = np.linspace(0, 150, 20)
y_range = np.linspace(0, 20, 20)
X, Y = np.meshgrid(x_range, y_range)
dH_dt = np.array([[dH_dt.subs({H: x_val, P: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dP_dt = np.array([[dP_dt.subs({H: x_val, P: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dH_dt, dP_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Hospedadores (H)')
plt.ylabel('Parásitos (P)')
plt.title('Retrato Fase del Modelo Nicholson-Bailey')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Nicholson-Bailey
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 4)
    for r in ic:
        for s in ic:
            x0 = [r * K, s]
            xs = odeint(func, x0, ts, args=(r, K, a, c, d))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Hospedadores (H)', fontsize=10)
    plt.ylabel('Parásitos (P)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 150)
    plt.ylim(0, 20)

    X, Y = np.mgrid[0:150:20j, 0:20:20j]
    u = r * X * (1 - X / K) - a * X * Y
    v = c * a * X * Y - d * Y
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

# Ejecutar la gráfica de trayectorias
grafica(nicholson_bailey)

```

##Interpretación

##Regiones de Atracción:

*__Equilibrio de Extinción (0, 0):__

 Si el retrato fase muestra una región de atracción alrededor del punto (0, 0), indica que para la mayoría de las condiciones iniciales, las poblaciones de parásitos y hospedadores tenderán a extinguirse a largo plazo. Esto puede ser interpretado como un escenario donde __el parásito no puede persistir en el ambiente debido a la baja tasa de contacto, la alta tasa de mortalidad o la baja capacidad de carga del medio ambiente para los hospedadores__.

* __Equilibrio de Exclusión Competitiva (K, 0):__

 Si el retrato fase muestra una región de atracción alrededor del punto (K, 0), indica que para la mayoría de las condiciones iniciales, la población de hospedadores se estabilizará en su capacidad de carga (K) mientras que la población de parásitos se extinguirá. Esto puede ser interpretado como un escenario donde __la respuesta inmune de los hospedadores o la competencia por recursos son lo suficientemente fuertes como para controlar la población de parásitos__.

##Coexistencia Estable:

Si el retrato fase muestra un punto de equilibrio estable dentro del cuadrante positivo del plano (H, P), indica que existe un estado de equilibrio donde tanto __la población de hospedadores como la de parásitos pueden persistir a largo plazo__. La ubicación de este punto de equilibrio y la forma de la región de atracción alrededor de él proporcionan información sobre las condiciones específicas que permiten la coexistencia estable.

##Ciclos:

Si el retrato fase muestra ciclos cerrados alrededor de un punto de equilibrio, indica que las poblaciones de parásitos y hospedadores __oscilan alrededor de un estado estable en un patrón cíclico__. El período y la amplitud de estos ciclos proporcionan información sobre la dinámica temporal de las poblaciones y los factores que influyen en sus fluctuaciones.
