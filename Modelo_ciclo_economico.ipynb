#Modelo de Goodwin de Ciclos Económicos ⏲


Entenderemos primero lo que es un ciclo económico de forma muy breve. Un ciclo económico es el __conjunto de fenómenos económicos dados en un periodo determinado de tiempo__, entre ellos se encuentra la crisis, depresión, reanimación y auge. Otra forma de verlos es como patrones recurrentes en la actividad económica de un país o región.

El Modelo de Goodwin de Ciclos Económicos es una teoría económica que intenta explicar la __dinámica de los ciclos económicos__, específicamente la __relación entre el desempleo y la inflación o la participación de los salarios en el ingreso total__.




##__¿Quién?__

 El modelo fue desarrollado por el economista británico __Richard Goodwin__ en la década de 1960 basado en ideas de la teoría marxista y la dinámica de depredador-presa de la biología. Goodwin fue un destacado economista conocido por sus contribuciones al análisis de los ciclos económicos y la dinámica económica.

La teoría económica __negaba la existencia de problemas en la demanda efectiva__, asumiendo que el sistema se mueve en un lugar equilibrado. La economía neoclásica se apoya en los viejos dogmas sin mucho cambio para describir la dinámica económica. Es ahí donde Goodwin entra en el campo para cambiar el paradigma, siendo uno de los economistas más originales y no conformistas, según lo dijo Pasinetti (1996), uno de sus discípulos.

Goodwin estaba convencido de que la evolución de la economía sólo podía entenderse mediante _sistemas de ecuaciones diferenciales no lineales_, pues decía que el capitalismo se encuentra con regularidad en un estado inestable, esto, según decía, es debido a que se trata de un juego jugado por millones de jugadores, quienes no pueden tener la suficiente información para jugarlo óptimamente, de esta forma, el sistema se bifurca en estados estables e inestables.


##__¿Qué es?__

 El modelo de Goodwin se basa en un sistema de ecuaciones diferenciales que __describen la interacción entre el empleo y la distribución del ingreso en una economía__. El modelo consta de dos ecuaciones:

* Una que relaciona el __cambio en la tasa de crecimiento del empleo con la tasa de crecimiento de la población laboral__ y __la tasa de explotación laboral__ (la relación entre los salarios reales y la productividad laboral).

Es decir, si la población laboral crece, se necesitan crear más empleos para matener la tasa de empleo constante.
Si la tasa de explotación laboral (productividad laboral vs. salarios) también está aumentando, los capitalistas requerirán menos trabajadores, pues obtienen más beneficios por cada trabajador.

* Otra que relaciona la __tasa de cambio de la participación de los beneficios__ con la __tasa de cambio de la tasa de explotación laboral__.

Es decir, si la tasa de explotación laboral aumenta (productividad laboral crece mas que los salarios reales), la participación de los beneficios en el ingreso total también aumenta, pues los capitalistas obtienen mayor valor de trabajo de los empleados sin alza de salarios.

##__¿Para qué?__

El propósito del modelo es proporcionar una explicación teórica de cómo los movimientos en la __distribución del ingreso entre beneficios y salarios pueden contribuir a la existencia de ciclos económicos__. Goodwin argumentó que existe una relación inversa entre la tasa de explotación laboral y el desempleo, lo que significa que a medida que disminuye la tasa de explotación laboral (es decir, los salarios reales aumentan en relación con la productividad), el desempleo tiende a disminuir. Sin embargo, esta disminución en la tasa de explotación laboral tiende a aumentar la inflación.

Se asume que los trabajadores gastan todo lo que ganan, y capitalistas invierten todo. El modelo asume que la __oferta siempre crea su propia demanda__, sin embargo, el mercado de trabajo no se vacía.

##**Ecuaciones del modelo**

Este modelo consiste en un sistema de ecuaciones diferenciales no lineales.

Supongamos que la productividad del trabajo crece a una tasa constante, expresado de la siguiente forma:

$$a=a_0 e^{\alpha t}$$

donde $\alpha=$ la tasa de crecimiento de la productividad

La tasa de crecimiento de la población es constante en el tiempo, así:

$$n=n_0e^{\beta t}$$

donde $\beta=$ la tasa de crecimiento de la fuerza de trabajo

Y la razón capital-producto $(\sigma)$ es constante, con $k$ el stock de capital y $q$ el nivel de producto, la ecuación que describe o anterior es la siguiente:

$$\sigma =\frac{k}{q}$$

 Supongamos que el empleo se representa por $l$ (demanda de trabajo) y $a$ la productividad media del trabajo, entonces tenemos la siguiente relación:

 $$l=\frac{q}{a}$$

 Los ingresos de los trabajadores se definen cono $u$ y $w$ la tasa de salarios:

 $$u=\frac{w l}{q}$$

 De la expresión del nivel de empleo tenemos:

  $$u=\frac{w}{a}$$

 Además, la tasa de empleo $(v)$, donde $l$ es la población ocupada y $n$ la oferta laboral:

 $$v=\frac{l}{n}$$

 Suponemos una relación de la tasa de crecimiento de los salarios reales y la tasa de empleo (versión lineal de la curva de Phillips), tenemos:

 $$\frac{\dot w }{w}=-\gamma +\rho v$$ con $\gamma>0$ y $\rho>0$

 Además del supuesto de que los beneficios se invierten totalmente, se escribe la tasa de acumulación como:

$$\dot k=(1-u)q$$

La expresión para la tasa de ganancia es:

$$r=\frac{\dot k}{k}=\frac{(1-u)q}{k}=\frac{(1-u)}{\sigma} $$

y dado que la razon capital- producto es constante, es facil mostrar que la tasa de crecimiento del producto es igual a la tasa de ganancia, así:
$$\frac{\dot q}{q}=\frac{\dot k}{k}=r$$

Las ecuaciones del sistema dinámico se obtienen de la diferenciación con respecto al tiempo de la participación de salarios en el ingreso ($u$) y de la tasa de empleo ($v$), así:

$$
\left\{
\begin{array}{ll}
\frac{\dot u}{u}=\frac{\dot w}{w}-\frac{\dot a}{a} & \\
\frac{\dot v}{v}=\frac{\dot l}{l}-\frac{\dot n}{n}
\end{array}
\right.
$$

En términos de las tasas de crecimiento de las variables de interés, de esta forma llegamos al modelo estándar del crecimiento:


$$
\left\{
\begin{array}{ll}
\dot u=[-(\gamma + \alpha)+\rho v]u & \\
\dot v=[\frac{(1 -u)}{\sigma}-(\alpha +\beta)]v
\end{array}
\right.
$$

La solución del sistema es una curva cerrada o ciclo limite

##Puntos Críticos

Para encontrar los puntos críticos del sistema, establecemos $(\dot u = 0)$ y $(\dot v = 0)$.

* Para $(\dot u = 0)$:

$$[-(\gamma + \alpha) + \rho v]u = 0$$

Esto se satisface si $u = 0$ o $[-(\gamma + \alpha) + \rho v] = 0$.

1. $u = 0$
2. $\rho v = \gamma + \alpha \implies v = \frac{\gamma + \alpha}{\rho}$

* Para $(\dot v = 0)$:

$\left[\frac{(1 - u)}{\sigma} - (\alpha + \beta)\right]v = 0$

Esto se satisface si $v = 0$ o $\left[\frac{(1 - u)}{\sigma} - (\alpha + \beta)\right] = 0$.

1. $v = 0$
2. $\frac{1 - u}{\sigma} = \alpha + \beta \implies u = 1 - \sigma(\alpha + \beta)$

Combinando estos resultados, obtenemos los puntos críticos del sistema:

1. $(u, v) = (0, 0)$
2. $(u, v) = \left(1 - \sigma(\alpha + \beta), \frac{\gamma + \alpha}{\rho}\right)$

##Clasificación de los Puntos Críticos

Para clasificar los puntos críticos, analizamos la matriz Jacobiana del sistema:

$$J(u, v) = \begin{pmatrix}
\frac{\partial \dot{u}}{\partial u} & \frac{\partial \dot{u}}{\partial v} \\
\frac{\partial \dot{v}}{\partial u} & \frac{\partial \dot{v}}{\partial v}
\end{pmatrix}$$

Calculamos las derivadas parciales:

$$\frac{\partial \dot{u}}{\partial u} = [-(\gamma + \alpha) + \rho v], \quad \frac{\partial \dot{u}}{\partial v} = \rho u
$$

$$
\frac{\partial \dot{v}}{\partial u} = -\frac{v}{\sigma}, \quad \frac{\partial \dot{v}}{\partial v} = \left[\frac{(1 - u)}{\sigma} - (\alpha + \beta)\right]
$$

Entonces, la matriz Jacobiana es:


$$J(u, v) = \begin{pmatrix}
-(\gamma + \alpha) + \rho v & \rho u \\
-\frac{v}{\sigma} & \frac{(1 - u)}{\sigma} - (\alpha + \beta)
\end{pmatrix}
$$

__Evaluación en los puntos críticos__

_En $(u, v) = (0, 0)$:_

$$
J(0, 0) = \begin{pmatrix}
-(\gamma + \alpha) & 0 \\
0 & \frac{1}{\sigma} - (\alpha + \beta)
\end{pmatrix}
$$

Los autovalores son $$\lambda_1 = -(\gamma + \alpha) \  y \ \lambda_2 = \frac{1}{\sigma} - (\alpha + \beta)$$.

- Si $\lambda_2 > 0$, este punto es un __punto silla__.
- Si $\lambda_2 < 0$, este punto es un __nodo estable__.

__En $(u, v) = \left(1 - \sigma(\alpha + \beta), \frac{\gamma + \alpha}{\rho}\right)$__

$$
J\left(1 - \sigma(\alpha + \beta), \frac{\gamma + \alpha}{\rho}\right) = \begin{pmatrix}
0 & \rho \left(1 - \sigma(\alpha + \beta)\right) \\
-\frac{\gamma + \alpha}{\sigma \rho} & 0
\end{pmatrix}
$$

Los autovalores son $$\lambda = \pm i\sqrt{\left(\frac{\gamma + \alpha}{\sigma \rho} \cdot \rho \left(1 - \sigma(\alpha + \beta)\right)\right)}$$.

Dado que los autovalores son puramente imaginarios, este punto es un __centro__.




```python
#Importamos las librerias necesarias
import sympy as sym #Para calculos simbolicos
import matplotlib.pyplot as plt #Para graficos
import numpy as np #Para calculos numericos
from scipy.integrate import odeint #Para integrar ecuaciones diferenciales ordinarias
import pylab as pl #Para graficos adicionales
```

```python
# Definir símbolos y parámetros
x, y = sym.symbols('x y')
a1, b1, a2, b2= 0.04, 0.02, 0.02, 0.04
# Definir las ecuaciones del modelo de Lotka-Volterra
dx_dt = x*(-a2+ b2 * y)
dy_dt = y*(a1-b1*x)
```

```python
# Encontrar los puntos críticos
critical_points = sym.solve([dx_dt, dy_dt], (x, y))

# Matriz Jacobiana
jacobi_matrix = sym.Matrix([[dx_dt.diff(var) for var in (x, y)],
                           [dy_dt.diff(var) for var in (x, y)]])
```

```python
# Función discriminante
def discriminant(J):
    return J.trace()**2 - 4 * J.det()
```

```python
# Función para determinar el tipo de punto fijo
def tipo(matrix):
    t = matrix.trace()
    d = matrix.det()
    disc = discriminant(matrix)

    if t > 0 and d > 0 and disc > 0:
        return "Región I (Nodo inestable)"
    elif t > 0 and d > 0 and disc < 0:
        return "Región II (Foco inestable)"
    elif t < 0 and d > 0 and disc < 0:
        return "Región III (Foco estable)"
    elif t < 0 and d > 0 and disc > 0:
        return "Región IV (Nodo estable)"
    elif d < 0 and disc < 0:
        return "Región V (Silla de montar)"
    else:
        return "Degenerado"

```

```python
# Clasificar la estabilidad del punto fijo
def clas(matrix):
    t = matrix.trace()
    d = matrix.det()

    if t < 0 and d > 0:
        return "El punto fijo es estable y atractivo"
    elif t <= 0 and d > 0:
        return "El punto fijo es estable"
    elif t > 0 and d < 0:
        return "El punto fijo es inestable"
    else:
        return "No tiene clasificación"
```

```python
# Evaluación en puntos críticos y clasificación
print("Puntos críticos y su clasificación:")
for point in critical_points:
    J = jacobi_matrix.subs({x: point[0], y: point[1]})
    tr = J.trace()
    det = J.det()
    disc = discriminant(J)

    print(f"\nPunto crítico: {point}")
    print(f"Traza: {tr}")
    print(f"Determinante: {det}")
    print(f"Discriminante: {disc}")
    print(f"Tipo: {tipo(J)}")
    print(f"Clasificación: {clas(J)}")
```

```python
# Retrato fase del modelo
def goodwin(X, t, a1, a2, b1, b2):
    x, y = X
    dx_dt =x*(-a2+ b2 * y)
    dy_dt = y*(a1-b1*x)
    return [dx_dt, dy_dt]

x_range = np.linspace(0, 10, 20)
y_range = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x_range, y_range)
dX_dt = np.array([[dx_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)
dY_dt = np.array([[dy_dt.subs({x: x_val, y: y_val}) for x_val in x_range] for y_val in y_range], dtype=float)

fig, ax = plt.subplots()
ax.streamplot(X, Y, dX_dt, dY_dt, density=1)
for point in critical_points:
    ax.plot(point[0], point[1], 'ro')
plt.xlabel('Tasa de empleo (x)')
plt.ylabel('tasa de salarios (y)')
plt.title('Retrato Fase del Modelo de Goodwin')
plt.grid()
plt.show()

# Graficar trayectorias del sistema de Goodwin
def grafica(func):
    ts = np.linspace(0, 50, 1000)
    ic = np.linspace(1, 5, 4)
    for r in ic:
        for s in ic:
            x0 = [r, s]
            xs = odeint(func, x0, ts, args=(a1, b1, a2, b2))
            plt.plot(xs[:, 0], xs[:, 1], "r-")

    plt.xlabel('Tasa de empleo (x)', fontsize=10)
    plt.ylabel('tasa de salarios (y)', fontsize=10)
    plt.tick_params(labelsize=10)
    plt.xlim(0, 10)
    plt.ylim(0, 10)

    X, Y = np.mgrid[0:10:10j, 0:10:10j]
    u= X*(-a2+ b2 * Y)
    v= Y*(a1-b1*X)
    plt.quiver(X, Y, u, v, color='b')
    plt.show()

    # Ejecutar la gráfica de trayectorias
grafica(goodwin)
```

## Observaciones:

* Al inicio, la tasa de empleo puede _aumentar_ debido a la acumulación de capital y una mayor demanda de trabajo.

Eventualmente, la tasa de empleo alcanza un _máximo_ y luego comienza a _disminuir_ debido al aumento de los salarios que reduce la rentabilidad y la inversión.

##Conclusiones

**Trayectoria Cíclica:**

La gráfica muestra una trayectoria cerrada en el espacio de fase, indicando que __el sistema tiene un comportamiento cíclico__. Esto significa que el sistema pasa por fases repetitivas de aumento y disminución en la tasa de empleo y la participación salarial.

**Interacción entre u y v:**

 La relación cíclica sugiere que __cuando la tasa de empleo es alta, la participación salarial también tiende a aumentar__, pero esto lleva a una disminución en la rentabilidad, lo que eventualmente reduce la tasa de empleo. A medida que la tasa de empleo baja, los salarios disminuyen, permitiendo una recuperación de la tasa de ganancia y un nuevo ciclo de aumento en la tasa de empleo.

**Punto de Equilibrio:**

 El centro del ciclo en la gráfica puede indicar un punto de equilibrio alrededor del cual las variables oscilan. Sin embargo, este equilibrio es inestable, y el sistema no permanece en este punto, sino que sigue oscilando en un ciclo continuo.

Este ciclo puede resumirse en las siguientes fases:

###Acumulación de Capital y Demanda de Trabajo:

Cuando la acumulación de __capital aumenta__, se __incrementa la demanda de trabajo__.
Esto conduce a un aumento en la tasa de empleo y en los salarios reales, que crecen más rápido que la productividad media del trabajo.


###Impacto en la Tasa de Ganancia:

Los salarios más altos __reducen la tasa de ganancia de los empresarios__.
Una menor tasa de ganancia reduce el ritmo de acumulación de capital y la tasa de crecimiento económico.


###Reducción de la Tasa de Empleo:

Eventualmente, la __tasa de empleo empieza a disminuir__.
Esto provoca un aumento de trabajadores desempleados.

###Disminución de los Salarios y Recuperación de la Tasa de Ganancia:

Con más desempleados, los __salarios comienzan a disminuir__.
Los salarios crecen a un ritmo menor que la productividad.
Esto permite que los __beneficios se recuperen y la tasa de ganancia aumente__.

###Inicio de un Nuevo Ciclo:

La recuperación de la tasa de ganancia lleva a un nuevo __ciclo de mayor acumulación de capital, mayor empleo y crecimiento económico__.

\

Las gráficas del retrato fase y la simulación del modelo de Goodwin proporcionan una _comprensión visual y cuantitativa de la dinámica cíclica_ entre la tasa de empleo y la participación salarial en una economía capitalista.

El ciclo económico explica cómo la acumulación de capital y el aumento de la demanda de trabajo inicialmente incrementan el empleo y los salarios. Sin embargo, los salarios altos eventualmente reducen la tasa de ganancia, desacelerando la acumulación de capital y el crecimiento económico. A medida que el empleo disminuye y los salarios caen, la tasa de ganancia se recupera, iniciando un nuevo ciclo de crecimiento económico.
