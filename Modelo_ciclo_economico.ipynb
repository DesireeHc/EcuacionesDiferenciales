#Modelo de Goodwin de Ciclos Económicos



Entenderemos primero lo que es un ciclo económico de forma muy breve. Un ciclo económico es el conjunto de fenómenos económicos dados en un periodo determinado de tiempo, entre ellos se encuentra la crisis, depresión, reanimación y auge. Otra forma de verlos es como patrones recurrentes en la actividad económica de un país o región.

El Modelo de Goodwin de Ciclos Económicos es una teoría económica que intenta explicar la dinámica de los ciclos económicos, específicamente la relación entre el desempleo y la inflación o la participación de los salarios en el ingreso total.




##__¿Quién?__

 El modelo fue desarrollado por el economista británico Richard Goodwin en la década de 1960 basado en ideas de la teoría marxista y la dinámica de depredador-presa de la biología. Goodwin fue un destacado economista conocido por sus contribuciones al análisis de los ciclos económicos y la dinámica económica.

La teoría económica negaba la existencia de problemas en la demanda efectiva, asumiendo que el sistema se mueve en un lugar equilibrado. La economía neoclásica se apoya en los viejos dogmas sin mucho cambio para describir la dinámica económica. Es ahí donde Goodwin entra en el campo para cambiar el paradigma, siendo uno de los economistas más originales y no conformistas, según lo dijo Pasinetti (1996), uno de sus discípulos.

Goodwin estaba convencido de que la evolución de la economía sólo podía entenderse mediante _sistemas de ecuaciones diferenciales no lineales_, pues decía que el capitalismo se encuentra con regularidad en un estado inestable, esto, según decía, es debido a que se trata de un juego jugado por millones de jugadores, quienes no pueden tener la suficiente información para jugarlo óptimamente, de esta forma, el sistema se bifurca en estados estables e inestables.


##__¿Qué es?__

 El modelo de Goodwin se basa en un sistema de ecuaciones diferenciales que describen la interacción entre el empleo y la distribución del ingreso en una economía. El modelo consta de dos ecuaciones:

Una que relaciona el cambio en la tasa de crecimiento del empleo con la tasa de crecimiento de la población laboral y la tasa de explotación laboral (la relación entre los salarios reales y la productividad laboral).
Otra que relaciona la tasa de cambio de la participación de los beneficios con la tasa de cambio de la tasa de explotación laboral.

##__¿Para qué?__

El propósito del modelo es proporcionar una explicación teórica de cómo los movimientos en la distribución del ingreso entre beneficios y salarios pueden contribuir a la existencia de ciclos económicos. Goodwin argumentó que existe una relación inversa entre la tasa de explotación laboral y el desempleo, lo que significa que a medida que disminuye la tasa de explotación laboral (es decir, los salarios reales aumentan en relación con la productividad), el desempleo tiende a disminuir. Sin embargo, esta disminución en la tasa de explotación laboral tiende a aumentar la inflación.

Las fluctuaciones económicas de este modelo se derivan del conflicto entre trabajadores y capitalistas, cuando, los salarios y las ganancias en el producto no se equilibran, la fuerza de negociación de trabajadores con la disciplina del mercado durante el desempleo, determina el ritmo en el que crecen los salarios, tasa de ganancias y crecimiento economico aunque a largo plazo no cambien los ingresos de asalariados y capitalistas.

Se asume que trabajadores gastan todo lo que ganan, y capitalistas invierten todo. El modelo asume que la oferta siempre crea su propia demanda, sin embargo, el mercado de trabajo no se vacía.


##**Ecuaciones del modelo**

Este modelo consiste en un sistema de ecuaciones diferenciales no lineales.

Supongamos que la productividad del trabajo crece a una tasa constante, expresado de la siguiente forma:

$$a=a_0 e^{\alpha t}$$

donde $\alpha=$ la tasa de crecimiento de la productividad

La tasa de crecimiento de la población es constante en el tiempo, así:

$$n=n_0e^{\beta t}$$

donde $\beta=$ la tasa de crecimiento de la fuerza de trabajo

Y la razón capital-producto $(\sigma)$ es constante, con $k$ el stock de capital y $q$ el nivel de producto, la ecuación que describe o anterior es la siguiente:

$$\sigma =\frac{k}{q}$$

 Supongamos que el empleo se representa por $l$ (demanda de trabajo) y $a$ la productividad media del trabajo, entonces tenemos la siguiente relación:

 $$l=\frac{q}{a}$$

 Los ingresos de los trabajadores se definen cono $u$ y $w$ la tasa de salarios:

 $$u=\frac{w l}{q}$$

 De la expresión del nivek de empleo tenemos:

  $$u=\frac{w}{a}$$

 Además, la tasa de empleo $(v)$, donde $l$ es la población ocupada y $n$ la oferta laboral:

 $$v=\frac{l}{n}$$

 Suponemos una relación de la tasa de crecimiento de los salarios reales y la tasa de empleo, tenemos:

 $$\frac{\dot w }{w}=-\gamma +\rho v$$ con $\gamma>0$ y $\rho>0$

```python
# Definimos las funciones f(x, y) y g(x, y)
a, b, c, d, e=1, 2, 3, 4, 5
def f(x, y):
    return x*(-(c+a)+d*y)
def g(x, y):
    return y*((1-x)/e-(a+b))
```

```python
# Definimos la función que describe el sistema de ecuaciones diferenciales
def dx_dt(x, t):
    return [f(x[0], x[1]), g(x[0], x[1])]

# Definimos las expresiones simbólicas para f(x, y) y g(x, y) usando SymPy
x, y = sym.symbols('x y')
f = x*(-(c+a)+d*y)
g = y*((1-x)/e-(a+b))
```

```python
# Encontramos los puntos fijos resolviendo f(x, y) = 0 y g(x, y) = 0
puntos_fijos = sym.solve([f_sym, g_sym], (x, y))
print("Puntos fijos:", puntos_fijos)
```

```python
   # Función para graficar las trayectorias
def grafica(dx_dt):
    ts = np.linspace(0, 4, 100)  # Generamos un vector de tiempo desde 0 hasta 4 con 100 puntos
    ic = np.linspace(-1, 1, 5)  # Generamos un vector de condiciones iniciales entre -1 y 1 con 5 puntos
    for r in ic:  # Iteramos sobre las condiciones iniciales en x
        for s in ic:  # Iteramos sobre las condiciones iniciales en y
            x0 = [r, s]  # Definimos la condición inicial
            xs = odeint(dx_dt, x0, ts)  # Resolvemos la ecuacion diferencial
            plt.plot(xs[:, 0], xs[:, 1], "r-")  # Graficamos la trayectoria en el plano
    plt.xlabel('x', fontsize=15)  # Etiqueta del eje x
    plt.ylabel('y', fontsize=15)  # Etiqueta del eje y
    plt.tick_params(labelsize=10)  # Tamaño de las etiquetas de los ejes
    plt.xlim(-10, 10)  # Limites del eje x
    plt.ylim(-10, 10)  # Limites del eje y

    X, Y = np.mgrid[-10:10:10j, -10:10:10j]  # Generamos una cuadricula de puntos
    u = f(X, Y)  # Calculamos el componente x del campo vectorial
    v = g(X, Y)  # Calculamos el componente y del campo vectorial
    pl.quiver(X, Y, u, v, color='b')  # Dibujamos el campo vectorial
    plt.show()  # Mostramos el grafico
```

```python
# Función para calcular el discriminante de la matriz
def D(A):
    return A.trace() ** 2 - 4 * A.det()  # Calculamos el discriminante usando la traza y el determinante
```

```python
# Función para determinar la región del punto fijo
def tipo(A, punto_fijo):
    t = A.trace().subs([(x, punto_fijo[0]), (y, punto_fijo[1])])  # Evaluamos la traza de la matriz en el punto fijo
    d = A.det().subs([(x, punto_fijo[0]), (y, punto_fijo[1])])  # Evaluamos el determinante de la matriz en el punto fijo
    grafica(dx_dt)  # Llamamos a la función para graficar las trayectorias
    discriminante = D(A).subs([(x, punto_fijo[0]), (y, punto_fijo[1])])  # Evaluamos el discriminante en el punto fijo
    if t > 0 and d > 0 and discriminante > 0:  # Condiciones para determinar la región basada en la traza, el determinante y el discriminante
        print("El punto fijo pertenece a la Región I")
    elif t > 0 and d > 0 and discriminante < 0:
        print("El punto fijo pertenece a la Región II")
    elif t < 0 and d > 0 and discriminante < 0:
        print("El punto fijo pertenece a la Región III")
    elif t < 0 and d > 0 and discriminante > 0:
        print("El punto fijo pertenece a la Región IV")
    elif d < 0 and discriminante < 0:
        print("El punto fijo pertenece a la Región V")
    else:
        print("El punto fijo es degenerado")
```

```python
# Función para clasificar la estabilidad del punto fijo
def clas(A, punto_fijo):
    t = A.trace().subs([(x, punto_fijo[0]), (y, punto_fijo[1])])
    d = A.det().subs([(x, punto_fijo[0]), (y, punto_fijo[1])])
    if t < 0 and d > 0:  # Condiciones para clasificar la estabilidad del punto fijo
        print("El punto fijo es estable y atractivo")
    elif t <= 0 and d > 0:
        print("El punto fijo es estable")
    elif t > 0 and d < 0:
        print("El punto fijo es inestable")
    else:
        print("No tiene clasificación")
```

```python
# Evaluamos las funciones para cada punto fijo encontrado
for punto_fijo in puntos_fijos:
    print("\nAnálisis del punto fijo:", punto_fijo)
    tipo(A, punto_fijo)
    clas(A, punto_fijo)
```
