# ¿Qué son los Sistemas Dinámicos?

## Introducción

¿Cómo predicen los economistas el comportamiento de los mercados financieros?

Los economistas utilizan __modelos dinámicos__ que __incorporan variables__ como tasas de interés, inflación y expectativas de los inversores. Aunque estos modelos pueden prever tendencias generales, la naturaleza compleja y a menudo caótica de los mercados financieros puede hacer que las predicciones precisas sean difíciles.

Este es sólo un ejemplo de las aplicaciones de los sistemas dinámicos en algunas otras áreas, pero para ello debemos conocer primero los conceptos básicos.

### Modelo

Un modelo es cualquier abstracción de la
realidad que guardamos en nuestra mente.

En el contexto de los sistemas dinámicos,un modelo es una __representación matemática___ que describe __cómo evoluciona un sistema a lo largo del tiempo__. Los modelos permiten __analizar, predecir y entender__ el comportamiento de sistemas complejos mediante el uso de ecuaciones y relaciones matemáticas.


### Sistema


Un sistema es un modelo matemático que cumple con ciertos
axiomas que lo caracterizan.

Dicho de otro modo, un sistema es un __conjunto de componentes interrelacionados que interactúan entre sí y evolucionan con el tiempo__. Estos componentes pueden ser variables físicas, químicas, biológicas, económicas, o de cualquier otro tipo, y las interacciones entre ellos se describen mediante leyes o ecuaciones matemáticas.

__Características__

* __Componentes y Variables de Estado:__

Los componentes del sistema son las __distintas partes que lo conforman__. Estas describen el estado actual del sistema. Por ejemplo, en un sistema ecológico, las variables de estado podrían ser las poblaciones de diferentes especies.

* __Interacciones y Relaciones:__

Las componentes del sistema interactúan entre sí de manera que __el estado de una componente puede influir en el estado de otras__. Estas interacciones se modelan a través de relaciones matemáticas o leyes físicas, como ecuaciones diferenciales o en diferencias.


* __Dinámica Temporal:__

Un sistema dinámico se caracteriza por __su evolución a lo largo del tiempo__. Esto significa que las variables de estado cambian con el tiempo según ciertas reglas.


* __Entradas y Salidas:__

Un sistema puede recibir entradas (inputs) del entorno y producir salidas (outputs). Las entradas pueden ser __variables externas que afectan al sistema__, mientras que las salidas son los __resultados del comportamiento del sistema__.


* __Fronteras del Sistema:__

Las fronteras definen __lo que está dentro del sistema y lo que está fuera__. Esto es importante para determinar qué factores se consideran en el análisis y cuáles se consideran externos.


* __Objetivo o Función del Sistema:__

Un sistema suele tener un __propósito o función__. Por ejemplo, un sistema económico puede tener como objetivo maximizar la producción, mientras que un sistema biológico puede buscar mantener el equilibrio entre especies.

## Sistemas dinámicos

La teoría de sistemas dinámicos es la base misma de casi cualquier tipo de modelos basados en reglas de sistemas complejos.

Ejemplos de sistemas dinámicos incluyen el __crecimiento de la población, un péndulo oscilante y los movimientos de los cuerpos celestes__.

La metodología de los sistemas dinámicos es una rama de las __matemáticas aplicadas y la teoría de control__ que se centra en el _análisis de sistemas que cambian con el tiempo_. Los sistemas dinámicos pueden ser descritos por ecuaciones diferenciales o ecuaciones en diferencias y pueden ser deterministas o estocásticos (cómo se modelan y caracterizan los sistemas en términos de su evolución temporal y la naturaleza de sus procesos).

Los pasos típicos involucrados en el estudio de sistemas dinámicos son los siguientes:



## Modelado del Sistema:

* __Identificación del sistema:__

 Definir el sistema de interés y determinar las variables relevantes que describen el estado del sistema.

* __Ecuaciones de estado:__

 Formular las ecuaciones que describen la evolución del sistema a lo largo del tiempo. Estas pueden ser ecuaciones diferenciales ordinarias (ODEs), ecuaciones en diferencias o sistemas de ecuaciones más complejos.

* __Condiciones iniciales:__ Establecer las condiciones iniciales del sistema, que son los valores de las variables de estado en el momento inicial.

### Análisis Cualitativo:

* __Estabilidad:__ Estudiar la estabilidad del sistema, es decir, cómo responde el sistema a perturbaciones. Esto puede implicar el análisis de puntos de equilibrio y su estabilidad.

* __Comportamiento a largo plazo:__ Investigar el comportamiento asintótico del sistema, como la existencia de ciclos límite, atractores extraños, o comportamiento caótico.

* __Diagramas de bifurcación:__ Analizar cómo cambian las soluciones del sistema cuando se varían ciertos parámetros, identificando transiciones entre diferentes tipos de comportamiento dinámico.

### Análisis Cuantitativo:

* __Solución numérica:__ Utilizar métodos numéricos para resolver las ecuaciones del sistema y obtener soluciones aproximadas cuando las soluciones exactas no son posibles.

* __Simulación:__ Implementar simulaciones computacionales para observar la evolución del sistema bajo diferentes condiciones y parámetros.

### Validación y Refinamiento:

* __Comparación con datos:__ Comparar las predicciones del modelo con datos experimentales o observacionales para validar su precisión.

* __Ajuste del modelo:__ Refinar el modelo en función de la comparación con los datos, ajustando parámetros o modificando las ecuaciones según sea necesario.


image.png


### Control y Optimización:

* __Diseño de controladores:__ Diseñar estrategias de control para influir en el comportamiento del sistema, logrando objetivos específicos como la estabilidad o la optimización de ciertos parámetros.

* __Evaluación de desempeño:__ Evaluar el desempeño del sistema bajo control y realizar ajustes para mejorar su eficiencia y eficacia.

Esta metodología se aplica en una amplia gama de campos, incluyendo la física, la biología, la economía, la ingeniería y muchas otras áreas donde los sistemas evolucionan con el tiempo.

Algunas conclusiones a las que podemos llegar sobre los sistemas dinámicos, son las siguientes:


### 1. Versatilidad en Modelado
Los sistemas dinámicos son __altamente versátiles__ y pueden ser utilizados para modelar una __amplia variedad de fenómenos en diferentes disciplinas__, desde la física y la ingeniería hasta la biología y las ciencias sociales. Esta versatilidad permite a los investigadores y profesionales describir y predecir el comportamiento de sistemas complejos de manera efectiva.

### 2. Diferentes Formas de Representación
Los sistemas dinámicos pueden ser __representados mediante ecuaciones diferenciales para sistemas continuos y ecuaciones en diferencias para sistemas discretos__. Esta capacidad de adaptación a diferentes tipos de problemas temporales amplía su aplicabilidad y utilidad en el análisis y diseño de sistemas.

### 3. Naturaleza Determinista vs. Estocástica
Los sistemas dinámicos pueden ser __deterministas__, donde el futuro del sistema está completamente determinado por sus condiciones iniciales y las leyes que lo gobiernan, o __estocásticos__, donde la aleatoriedad y la incertidumbre juegan un papel crucial. Esta distinción es fundamental para entender y predecir el comportamiento de sistemas en presencia de incertidumbre.

### 4. Importancia de las Condiciones Iniciales
En sistemas deterministas, las condiciones iniciales son cruciales para la __predicción precisa del comportamiento futuro del sistema__. Incluso pequeñas variaciones en las condiciones iniciales pueden llevar a resultados significativamente diferentes, un fenómeno conocido como sensibilidad a las condiciones iniciales, común en sistemas caóticos.

### 5. Aplicaciones en el Mundo Real
Los modelos de sistemas dinámicos tienen __aplicaciones prácticas en numerosos campos__. En ingeniería, se utilizan para diseñar y controlar sistemas mecánicos, eléctricos y de control. En biología, modelan el crecimiento de poblaciones y la propagación de enfermedades. En economía y finanzas, ayudan a entender y predecir comportamientos del mercado.

### 6. Simulación y Análisis Numérico
Debido a la complejidad de muchos sistemas dinámicos, a menudo es necesario recurrir a __simulaciones y métodos numéricos para analizar su comportamiento__. Estas herramientas permiten explorar escenarios complejos y obtener información valiosa que no sería posible derivar analíticamente.

### 7. Interdisciplinariedad
El estudio de sistemas dinámicos es __inherentemente interdisciplinario__. Requiere conocimientos de matemáticas, física, informática y la disciplina específica del problema que se está abordando. Esta interdisciplinariedad fomenta la colaboración y el intercambio de conocimientos entre diferentes campos.

### 8. Potencial Predictivo
Cuando se modelan y analizan correctamente, los sistemas dinámicos pueden proporcionar poderosas __predicciones sobre el comportamiento futuro de un sistema__. Esto es particularmente valioso en áreas como la meteorología, la planificación de recursos, y la gestión de sistemas complejos como redes eléctricas y transporte.

### 9. Desafíos en la Modelización
A pesar de sus beneficios, modelar sistemas dinámicos también presenta desafíos significativos. Capturar todas las variables relevantes, manejar la complejidad computacional y lidiar con la incertidumbre y el ruido son algunos de los retos comunes en el modelado de sistemas dinámicos.

### 10. Importancia del Tiempo
El aspecto temporal es fundamental en los sistemas dinámicos. Comprender cómo las __variables cambian y evolucionan con el tiempo__ es crucial para el análisis y la predicción del comportamiento del sistema. Esto incluye tanto la dinámica a corto plazo como las tendencias a largo plazo.

En conclusión, los sistemas dinámicos son herramientas esenciales y poderosas para comprender, modelar y predecir el comportamiento de sistemas complejos en una variedad de campos. Su capacidad para adaptarse a diferentes tipos de problemas y su aplicación práctica en el mundo real los hacen indispensables en la ciencia y la ingeniería modernas.
