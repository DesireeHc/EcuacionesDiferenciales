```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def discriminante(a, b, c):
    discriminante = b**2 - 4 * a * c
    return discriminante
```

```python
def generar_matrices(n, min_valor, max_valor):
    matrices = []
    for _ in range(n):
        a = np.random.randint(min_valor, max_valor + 1)
        b = np.random.randint(min_valor, max_valor + 1)
        c = np.random.randint(min_valor, max_valor + 1)
        matriz = np.array([[a, b], [0, c]])  # Matriz 2x2 con c en la diagonal inferior derecha
        matrices.append(matriz)
    return matrices
```

```python
n_matrices = 1000  # Número de matrices a generar
min_valor = -10  # Valor mínimo para elementos aleatorios
max_valor = 10  # Valor máximo para elementos aleatorios

matrices = generar_matrices(n_matrices, min_valor, max_valor)
```

```python
matrices_discriminante_cero = []
for matriz in matrices:
    if discriminante(matriz[0, 0], matriz[0, 1], matriz[1, 0]) == 0:
        matrices_discriminante_cero.append(matriz)
```

```python
a_discriminante_cero = []
c_discriminante_cero = []

for matriz in matrices_discriminante_cero:
    a_discriminante_cero.append(matriz[0, 0])
    c_discriminante_cero.append(matriz[1, 0])
```

```python
plt.figure(figsize=(10, 6))

# Gráfica de puntos para a vs c
plt.scatter(a_discriminante_cero, c_discriminante_cero, color='blue', label='a vs c (discriminante = 0)')

# Ejes y etiquetas
plt.xlabel('a')
plt.ylabel('c')
plt.title('Relación entre a y c para discriminante = 0')
plt.grid(True)

# Limitar rango de ejes
plt.xlim([min_valor, max_valor])
plt.ylim([min_valor, max_valor])

# Leyenda
plt.legend()

# Mostrar gráfica
plt.show()
```

```python
def generar_matrices_aleatorias(n, min_valor, max_valor):
    matrices = []
    for _ in range(n):
        matriz = np.random.randint(min_valor, max_valor + 1, size=(2, 2))
        matrices.append(matriz)
    return matrices

def calcular_traza(matriz):
    return np.trace(matriz)

def calcular_determinante(matriz):
    return np.linalg.det(matriz)

def calcular_discriminante(matriz):
    b = matriz[0][1]
    a = matriz[0][0]
    c = matriz[1][0]

    discriminante = (b ** 2) - (4 * a * c)
    return discriminante

```

```python
n_matrices = 1000  # Número de matrices a generar
min_valor = -10  # Valor mínimo para elementos aleatorios
max_valor = 10  # Valor máximo para elementos aleatorios

matrices = generar_matrices_aleatorias(n_matrices, min_valor, max_valor)

traces = []
determinantes = []
discriminantes = []

for matriz in matrices:
    trace = calcular_traza(matriz)
    determinante = calcular_determinante(matriz)
    discriminante = calcular_discriminante(matriz)

    traces.append(trace)
    determinantes.append(determinante)
    discriminantes.append(discriminante)
```

```python
matrices_discriminante_cero = []
traces_discriminante_cero = []
determinantes_discriminante_cero = []

for i, discriminante in enumerate(discriminantes):
    if discriminante == 0:
        matrices_discriminante_cero.append(matrices[i])
        traces_discriminante_cero.append(traces[i])
        determinantes_discriminante_cero.append(determinantes[i])
```

```python
plt.figure(figsize=(10, 6))

# Gráfica de puntos para matrices con discriminante = 0
plt.scatter(traces_discriminante_cero, determinantes_discriminante_cero, color='red', label='Discriminante = 0')

# Eje x
plt.xlabel('Traza')

# Eje y
plt.ylabel('Determinante')

# Título de la gráfica
plt.title('Plano Traza-Determinante con Discriminante = 0')

# Cuadrícula
plt.grid(True)

# Leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
